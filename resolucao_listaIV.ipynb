Questão 1

```python
#a) Adicione a chave "custo_total" a cada dicionário de pedidos, calculada como quantidade × preco_unit.
for item in pedidos:
    item['custo_total'] = round(item['quantidade']*item['preco_unit'],2)
print(pedidos)
```

```python
#b) Crie a lista ranking com os pedidos ordenados do maior para o menor custo_total. Use sorted() com lambda.
ordenado = sorted(pedidos, key=lambda item: item['custo_total'],reverse=True)
print(ordenado)
```

Questão 2

```python
# a) Crie o dicionário gastos com o gasto total (soma de custo_total) por obra. Percorra pedidos com for. Atenção: custo_total deve ter sido calculado na Questão 1.
gastos = {}
for item in pedidos:
    if item['obra'] in gastos.keys():
        gastos[item['obra']] += item['custo_total']
    else:
        gastos[item['obra']] = item['custo_total']

print(gastos)
```

```python
#b) Crie o dicionário num_pedidos com o total de pedidos realizados por cada obra.
num_pedidos = {}
for item in pedidos:
    if item['obra'] in num_pedidos.keys():
        num_pedidos[item['obra']] += 1
    else:
        num_pedidos[item['obra']] = 1

print(num_pedidos)
```

```python
#c) Use max() com lambda sobre gastos.items() para encontrar a obra com maior gasto total. Exiba nome e valor (round 2).
maior_valor = max(gastos.items(), key=lambda x: x[1])
print(maior_valor)
```

```python
# a) Crie o dicionário por_material com a quantidade total solicitada (soma de quantidade) de cada material.
por_material = {}
for item in pedidos:
    if item['material'] in por_material.keys():
        por_material[item['material']] += 1
    else:
        por_material[item['material']] = 1

print(por_material)
```

```python
# b) Use max() com lambda sobre por_material.items() para exibir o material com maior quantidade total solicitada.
maior_quantidade = max(por_material.items(), key=lambda x: x[1])
print(maior_quantidade)
```

```python
# c - Crie a lista ranking_materiais ordenando por_material.items() do maior para o menor volume com sorted() + lambda. Exiba o resultado.

ranking_material = sorted(por_material.items(), key=lambda x: x[1], reverse=True)
print(ranking_material)
```

Questão 5

```python
#a) Adicione a chave "iue" a cada dicionário de pedidos, calculando o IUE conforme a fórmula. Use round() com 2 casas decimais. (custo_total deve existir da Questão 1.)

for item in pedidos:
    if item['prazo_dias'] > 0:
        item['iue'] = round(item['custo_total']/item['prazo_dias'],2)
print(pedidos)
```

```python
#b) Use max() com lambda para encontrar o pedido com o maior IUE e armazene-o na variável critico.

critico = max(pedidos, key=lambda x: x['iue'])
print(critico)
```
