**ESTRUTURAS DE DECISÃO**

**if**

***Quando precisamos fazer um teste simples sem novas verificações***

***Caso não seja verdadeiro, simplesmente pula e vida que segue.***

```python
#if sozinho
#if verdade: alguma coisa
x = 0
if x == 0:
    print('É zero')
#Não tem mais nada, se x fosse 1 só pulava e fazia nada.
x = 1
if x == 0:
    print('É zero')
```

**if/else**

if - primeira opção

else - opção que não atende a primeira

***Usado apenas quando temos absolutamente 2 opções exclusivas (Ex.: Par ou Ímpar, Verdadeiro ou Falso, ...)***

***Só vai acontecer uma delas!!!***

```python
#if com else
x = 0
if x == 0:
    print("X é zero")
else:#Não tem argumentos
    print('X não é zero')
#Nesse caso se x fosse 1 teria a resposta 'x não é zero'
```

**if/elif (sem else)**

if - a primeira opção

elif - as demais opções com condições extras

***Usado apenas quando temos mais de 2 opções exclusivas (com verdades definidas). Ex.: Fórmula de área de triângulos***

***Só vai acontecer uma delas!!!***



```python
#Agora x assumiu o valor 1
x = 1
if x == 0:
    print("X é zero")
else:#Não tem argumentos
    print('X não é zero')
```

**if/elif/else**

if - a primeira opção

elif - as demais

else - a que restar por último

***Testa o if, caso não seja verdade, testa o(s) elif, caso nenhuma seja verdade executa o else***


***Caso encontre alguma verdade no if, no elif, antes do else, ele para e pula tudo até o final. Só vai acontecer uma delas!!!***

```python
#Se x igual a 10
x = 10
if x > 10: #Falso
    print('X é maior que 10')
elif x < 10: #Falso
    print('X é menor que 10')
else:#Verdadeiro, pois nada anteriormente atendeu
    print('x é igual a 10')

#else sempre vai ser o contrário de tudo que vier antes, caso de excessão
```

**MATCH/CASE** - Quando precisamos de muitas seleções fixas

* cada CASE deve ser único e determinado por um valor numérico ou texto, não podemos usar expressões lógicas como no if/elif/else

* só executa uma opção por vez

* caso não seja lido uma opção válida, aciona o valor default -> case _:

```python
#Match/case com números
x = int(input('Digite um valor de 1 a 4: '))
match(x):
    case 1:
        print('Opção 1 (funciona como um if, mas com valor fixo)')
    case 2:
        print('Opção 2 (funciona como um elif, mas com valor fixo)')
    case 3:
        print('Opção 3 (funciona como um elif, mas com valor fixo)')
    case 4:
        print('Opção 4 (funciona como um elif, mas com valor fixo)')
    case _:
        print('default (funciona como um else dos cases anteriores)')
```

```python
#Match/case com texto
x = input('Digite a sua opção (A - D): ').lower() #deixa minusculas (ou maiusculas -> .upper()) para que seja detectado exatamente o escrito nas opções
match(x):
    case 'a': #se aqui está minúscula usar lower() / se está aqui maiúscula usar upper()
        print('Opção A (funciona como um if, mas com valor fixo)')
    case 'b':
        print('Opção B (funciona como um elif, mas com valor fixo)')
    case 'c':
        print('Opção C (funciona como um elif, mas com valor fixo)')
    case 'd':
        print('Opção D (funciona como um elif, mas com valor fixo)')
    case _:
        print('default (funciona como um else dos cases anteriores)')
```

**ESTRUTURAS DE REPETIÇÃO**

**Estrutura de repetição pré-fixada com 1 argumento:**

***for i in range(fim)***

***//execuções***


Leitura:***para*** um valor i (iniciando de 0) ***vá até*** fim-1

```python
print('For in range com 1 parametro:',end = " ")
for i in range(10): #0 a 9 ->  limite do range(elemento-1)
    print(i,end = " ")
```

**Estrutura de repetição pré-fixada com 2 argumentos:**

***for i in range(inicio, fim)***

***//execuções***


Leitura:***para*** um valor i (valor de inicio) ***vá até*** até o fim-1

```python
print('For in range com 2 parametros:',end = " ")
for i in range(1,10): #1 a 9 -> inicio = 1 e limite do range(elemento-1) = 9
    print(i,end = " ")
```

**Estrutura de repetição pré-fixada com 3 argumentos (crescente):**

***for i in range(inicio, fim, passo(positivo -> crescente)***

***//execuções***


Leitura:***para*** um valor i (valor de inicio) ***vá até*** até o fim-1 ***pulando de*** passo ***em*** passo.

```python
print('For in range com 3 parametros:',end = " ")
for i in range(1,10,2): #1 a 9 -> inicio = 1 e limite do range(elemento(10)-1) = 9
#porém aqui vai pular de 2 em 2 até o limite
    print(i,end = " ")
```

**Estrutura de repetição pré-fixada com 3 argumentos (decrescente):**

***for i in range(inicio (maior valor), fim (menor valor), passo reverso (negativo -> decrescente)***

***//execuções***


Leitura:***para*** um valor i (valor de inicio) ***vá descrescentemente até*** até o fim+1 ***pulando de*** passo ***em*** passo.

```python
print('For in range com 3 parametros:',end = " ")
for i in range(10,1,-1): #10 a 2 -> inicio = 10 e limite do range(elemento(1) +1) = 2
#porém aqui vai pular de 2 em 2 até o limite
    print(i,end = " ")
```

**Estrutura de repetição pós-fixada:**

***while argumento:***

***//execuções***

***//incremento ou ponto de parada***

```python
#Quando usar o While? - quando o limite de repetição é pós-fixado (varia com o problema)

x = 0
while True: #Enquanto for verdade. Limite não determinado (True ou False)
    print(x, end = " ") #imprime o valor de x. (end = " ") serve para manter as respostas na mesma linha
    x+=1 #x = x + 1 #incrementa de x em +1.
    if x > 10: #verifica se ele ainda é válido para o while.
        break #caso não seja mais verdade, para o loop.-> False

print('\n')
x = 0
```

```python
#Quando não usar o While? - quando o limite de repetição é fixado (não varia com o problema)

while x < 50:#Enquanto for verdade. limite determinado (melhor usar for)
    print(x, end = " ")
    x+=10

# que é exatamente igual a:
print('\nQue é igual a:')
for i in range(0,50,10):
    print(i,end = " ")

#Então nesses casos não vale a pena usar while...
#Usar While True: ou quando for valor que tenha verificações mais complexas que apenas verificar o valor.
```
