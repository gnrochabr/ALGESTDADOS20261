ms = [80100, 81700, 79500, 83000]

Ele quer aplicar a transformação em todos os elementos de uma vez, sem escrever um laço for explícito, usando uma abordagem funcional.

Escreva o código usando map() combinado com lambda para gerar uma lista com os tempos em segundos. Explique o que map() retorna diretamente (antes da conversão) e por que é necessário envolvê-lo com list(). Qual seria o resultado final esperado para essa lista?

```python
ms = [80100, 81700, 79500, 83000]
mapeado = list(map(lambda x: x/1000,ms))
print(mapeado)
```

O que map() retorna diretamente: Antes da conversão, a função map() retorna um objeto iterável (especificamente, um <map object> localizado na memória). Ele é um iterador preguiçoso (lazy), o que significa que ele não calcula todos os valores de uma vez para armazenar em uma lista, mas sim sob demanda.Por que é necessário envolvê-lo com list(): Como o map() retorna apenas uma referência para um gerador de passos, o Python não exibe os valores diretamente. Envolvê-lo com list() força o iterador a calcular e avaliar todos os elementos imediatamente, agrupando e formatando o resultado final como uma estrutura de lista padrão.

10 - Ao final da corrida, o analista precisa calcular o tempo total percorrido pelo piloto somando todos os tempos de volta registrados. Os dados estão disponíveis na seguinte lista:

from functools import reduce
voltas = [80.1, 79.5, 80.8, 81.7]

Em vez de usar sum() ou um laço acumulador, o analista quer demonstrar o uso da função reduce() da biblioteca functools.

Escreva a chamada correta usando reduce() combinado com lambda e calcule o resultado esperado. Em seguida, descreva passo a passo como reduce() processa a lista internamente — isto é, quais são os valores de x e y em cada aplicação da lambda. Por que a lambda precisa obrigatoriamente de dois parâmetros nesse caso?

```python
from functools import reduce
voltas = [80.1, 79.5, 80.8, 81.7]
tempo_total = reduce(lambda x,y:x+y,voltas)
#Pega os valores da lista e reduz a um valor único
print(tempo_total)
```

A função reduce() requer obrigatoriamente dois parâmetros na função lambda porque ela funciona aplicando uma operação binária (em pares) consecutiva para reduzir uma lista a um único valor. O primeiro parâmetro atua como um acumulador (resultado temporário) e o segundo representa o próximo elemento da lista. [1, 2, 3, 4]
Para demonstrar o passo a passo interno, vamos utilizar um exemplo padrão de soma:
reduce(lambda x, y: x + y, [1, 2, 3, 4]) [5]
## Passo a Passo da Execução Interna
Se nenhum valor inicial for fornecido, o reduce() extrai os dois primeiros elementos da lista para iniciar o processo:

| Etapa [6] | Valor de x (Acumulador) | Valor de y (Elemento Atual) | Operação (x + y) | Resultado (Novo x) |
|---|---|---|---|---|
| 1ª chamada | 1 (1º item da lista) | 2 (2º item da lista) | 1 + 2 | 3 |
| 2ª chamada | 3 (Resultado anterior) | 3 (3º item da lista) | 3 + 3 | 6 |
| 3ª chamada | 6 (Resultado anterior) | 4 (4º item da lista) | 6 + 4 | 10 (Resultado final) |

Nota: Se você passar um valor inicial, ex: reduce(lambda x, y: x + y, [1, 2, 3, 4], 10), a primeira chamada usará x = 10 (o valor inicial) e y = 1 (o primeiro item da lista).
## Por que a lambda precisa obrigatoriamente de dois parâmetros?

* Papéis distintos: O parâmetro x guarda a memória do que já foi processado (o acumulado), enquanto o y introduz o novo dado a ser processado.
* Redução gradual: Computadores só conseguem processar funções matemáticas binárias com dois argumentos por vez. A estrutura de dois parâmetros permite que o reduce() "varra" a lista combinando o passado (x) com o presente (y) até que reste apenas um elemento.
* Erro de assinatura: Se a sua lambda recebesse apenas um parâmetro, o reduce() quebraria ao tentar injetar dois valores nela, gerando um erro de tipo (TypeError).
