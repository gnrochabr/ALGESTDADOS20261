```python
# 01 Contador com While
# Usando while, exiba os números de 1 até 10, um por linha.
#Use uma variável de controle iniciada em 1 e incremente a cada iteração.

contador = 1
while contador<=10:
    print(contador)
```

```python
#02 Senha Segura
#Defina uma senha correta no código.
#Usando while, peça a senha ao usuário repetidamente até que ele acerte.
#A cada tentativa errada, exiba 'Senha incorreta, tente novamente.'.
#Ao acertar, exiba 'Acesso liberado!' e quantas tentativas foram necessárias.

contador = 0
senha = 'python2026'

while True:
    senha_digitada = input('Digite a senha: ')
    contador+=1
    if senha_digitada == senha:
        print(f'Acesso liberado! Usou {contador} tentativas!')
        break
    else:
        print('Senha incorreta, tente novamente.')
```

```python
#03 Adivinhe o Número
#Defina um número secreto entre 1 e 100 no código.
#Usando while, leia tentativas do usuário até acertar.
#A cada erro, diga se o número secreto é maior ou menor.
#Ao final, exiba quantas tentativas foram usadas.
#import random
#numero = random.randint(1,100)
numero = 13
cont = 0
while True:
    tentativa = int(input('Digite sua tentativa (1 - 100): '))
    cont+=1
    if 1 <= tentativa <= 100:
        if tentativa == numero:
            print(f'Acertou Miseravi! Usando {cont} tentativas!')
            break
        elif tentativa > numero:
            print('O número é menor')
        else:
            print('O número é maior')
    else:
        print('Os números tem que estar entre 1 e 100')
```

```python
#04 Soma até Estourar
#Leia um limite positivo. Usando while, leia números inteiros positivos do usuário e some-os.
#Enquanto a soma não ultrapassar o limite, continue pedindo números.
#Exiba a soma final e a quantidade de números lidos.

limite = 100 #Define o limite
soma = 0 #Inicializa a soma
cont = 0 #Inicializa o contador
while soma <= limite: #Verifica se estorou o limite
    #Caso continue sendo Falso:
    num = int(input('Digite um número inteiro positivo:')) #Pede uma nova tentativa
    cont+=1 #Cada tentativa é contado como mais uma
    soma+=num #Soma é atualizada: soma = soma + num
#Caso tenha estourado
print(f'A Soma foi {soma} e estorou na {cont}ª tentativa!')

```

```python
#5
#Usando while, leia uma nota do usuário repetidamente até que ela seja válida (entre 0 e 10).
#A cada entrada inválida, exiba uma mensagem de erro específica.
#Ao obter um valor válido, exiba 'Nota registrada: X'.

while True:
    nota = float(input('Digite uma nota válida (0 - 10): '))
    if 0<= nota <=10:
        print(f'Nota registrada: {nota}')
        break
    else:
        print('Valor inválido - Tente novamente com valor entre 0 e 10')
```

```python
#6
#Leia um número inteiro positivo N.
contagem = int(input('Número inicial: '))
if contagem > 0:
    #Usando while, faça a contagem regressiva de N até 0, exibindo cada número. Ao chegar em 0, exiba 'Fim!'.
    limite = 0
    while contagem > limite:
        print(contagem)
        contagem-=1
    print('Fim!')
else:
    print('Número inválido')
```

```python
# 7
# Usando while, leia números do usuário até que ele digite 0.
# Ignore o 0 na soma.
# Ao final, calcule e exiba a média de todos os números digitados (exceto o 0) com 2 casas decimais.
# Trate o caso de nenhum número válido ser digitado.
soma = 0
contador = 0
while True:
    numero = int(input('Digite um número:'))
    if numero == 0:
        break
    elif numero < 0:
        print('Número inválido')
    else:
        soma+=numero
        contador+=1

media = soma/contador
print(f'Média: {media:.2f}')
```

```python
#9
#Sequência de Fibonacci
#Leia um número N positivo.
#Usando while, gere e exiba os N primeiros termos da sequência de Fibonacci (0, 1, 1, 2, 3, 5...).
#Use variáveis descritivas como termo_anterior, termo_atual e proximo_termo.

numero =  int(input("Digite um valor inteiro positivo:"))
if numero > 0:
    termo_anterior = 0
    termo_atual = 1
    proximo_termo = 1
    while numero > 0:
        print(termo_atual)
        proximo_termo = termo_atual + termo_anterior
        termo_anterior = termo_atual
        termo_atual = proximo_termo
        numero-=1

```

```python
#10 Usando while, leia números inteiros do usuário até que ele digite 0.
#Controle a sequência atual de positivos consecutivos e a maior sequência encontrada.
#Ao final, exiba o tamanho da maior sequência de números positivos consecutivos.
contador = 0
maior = 0
while True:
    sequencia = int(input('Digite um número inteiro (!=0):'))
    if sequencia != 0:
        contador+=1
    else:
        if contador > maior:
            maior = contador
        print(f'Fim da Sequencia! - Sua Maior sequência é {maior}')
        contador = 0
        while True:
            letra = input('Deseja continuar? S/N: ')
            if letra.upper() == 'S':
                print('Nova Sequência:')
                break
            elif letra.upper() == 'N':
                break
            else:
                print('Opção Inválida')
        if letra.upper() == 'N':
            print(f'Fim de Jogo! - Sua Maior sequência foi {maior}')
            print('Xau!')
            break
```

```python
#11
#Usando while, exiba um menu que permite ao usuário escolher um número para ver sua tabuada (1-10)
#ou digitar 0 para sair. Para cada número escolhido, calcule e exiba a tabuada completa.
#Valide que a escolha está entre 0 e 10.

while True:
    numero = int(input('\nDigite a tabuada desejada: '))
    if 0 <= numero <= 10:
        if numero == 0:
            print('Vazando...')
            break
        indice = 0
        print(f'\n-- Tabuada de {numero} --')
        while indice <=10:
            print(f'{numero} x {indice} = {numero*indice}')
            indice+=1

    else:
        print('Você digitou fora da escala de 0 a 10')

```

```python
#12
#Leia uma palavra.
#Usando while com um índice de controle, compare o primeiro com o último caractere,
#o segundo com o penúltimo, e assim por diante, para verificar se a palavra é palíndromo.
#Exiba o resultado. Dica: use len() e indexação de strings.

palavra = input('Digite um palavra: ')
inicio = 0
fim = len(palavra)-1
meio = fim/2
while True:
    if palavra[inicio].upper()!=palavra[fim].upper():
        print('Não é palindromo')
        break
    elif inicio == meio or fim == meio:
        print(f'A {inicio+1}ª letra é {palavra[inicio]}')
        print('É palíndromo')
        break
    else:
        print(f'A {inicio+1}ª letra é {palavra[inicio]} e A {fim+1}ª letra é {palavra[fim]}')
        inicio+=1
        fim-=1

```

```python
#13 - Usando while, simule uma caixa registradora:
#leia o nome e preço de produtos até o usuário digitar 'fim' como nome.
#Valide que o preço é positivo. Ao final, exiba a quantidade de produtos,
#o subtotal, o imposto (10%) e o total a pagar.
compra = 0
while True:
    produto = input('Digite o nome do produto:')
    if produto.lower() != 'fim':
        preco = float(input('Digite o valor do produto: '))
        quantidade = int(input('Digite a quantidade: '))
        subtotal = preco*quantidade
        compra+= subtotal
        print(f'\n{produto} - R${preco:.2f} - {quantidade} - R${subtotal:.2f}\n')
    else:
        print(f'\nTotal da compra: R$ {compra:.2f}')
        print(f'Impostos: R$ {compra*0.1:.2f}')
        print(f'Valor Final: R$ {compra*1.1:.2f}')
        break
```

```python
#14
#Leia um valor inicial, uma razão (valor a somar a cada passo) e um limite, todos positivos.
#Usando while, exiba os termos da progressão aritmética enquanto o valor atual for menor ou igual ao limite.
#Exiba também a quantidade de termos gerados.

inicio = int(input('Inicio: '))
razao = int(input('Progressão: '))
limite = int(input('Limite: '))
contador = 0

if limite > 0 and razao > 0 and inicio > 0:
    print(inicio)
    while inicio + razao < limite:
        termo = inicio + razao
        inicio = termo
        print(termo)
```
