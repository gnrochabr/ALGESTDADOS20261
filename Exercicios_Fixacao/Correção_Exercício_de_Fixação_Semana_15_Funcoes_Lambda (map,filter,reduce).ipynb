```python
''' 1 - Em programação funcional, funções lambda são expressões anônimas de uso único. Crie uma função lambda que receba um número inteiro fornecido pelo usuário e retorne o seu quadrado. Armazene a lambda em uma variável chamada quadrado, aplique-a ao valor digitado e exiba o resultado com uma mensagem descritiva.'''
valor = int(input('Valor: '))
quadrado = lambda x: x**2
print(f'O quadrado de {valor} é {quadrado(valor)}')
```

```python
''' 2 - Lambdas permitem encapsular verificações lógicas de forma compacta. Crie uma função lambda chamada eh_par que receba um inteiro e retorne True se ele for divisível por 2 e False caso contrário. Solicite um número ao usuário, aplique a lambda e exiba a classificação par ou ímpar com base no retorno.'''
valor = int(input('Valor: '))
eh_par = lambda x: x%2==0
#res = lambda x: 'par' if eh_par(x) else 'ímpar'
print(eh_par(valor))
```

```python
''' 3 - A função map() aplica uma transformação a todos os elementos de um iterável sem laços explícitos. Solicite ao usuário 6 números inteiros e armazene-os em uma lista. Utilize map() com uma lambda para gerar uma nova lista com o dobro de cada valor. Exiba a lista original e a transformada lado a lado.'''
lista = []
for _ in range(6):
    valor = int(input('Valor: '))
    lista.append(valor)
#lista = [int(input('Valor: ')) for _ in range(6)] #List_comprehension
nova_lista = list(map(lambda x: x*2, lista))
print(lista, nova_lista)
```

```python
'''4 - Em sistemas de monitoramento internacional, conversões de unidade são aplicadas a conjuntos de dados. Dado uma lista de temperaturas em Celsius fornecida pelo usuário, utilize map() com a fórmula F = C × 9/5 + 32 para converter toda a lista para Fahrenheit. Exiba os pares de conversão.'''

temperaturas = [36.5,22.4,21.2,40.2,12.4]
print('Fahrenreit:',list(map(lambda x: round(x * (9/5) + 32,2), temperaturas)))
```

```python
''' 5 - Em sistemas de padronização de dados textuais, é comum converter strings para um formato uniforme. Peça ao usuário 5 nomes quaisquer e utilize map() com uma lambda para gerar uma nova lista com todos os nomes em letras maiúsculas. Exiba a lista original e a normalizada, destacando as diferenças.'''

nomes = ["Ana", "Bruno", "Carla", "Diego", "Elena", "Felipe"]
print(nomes,'->',list(map(lambda x: x.upper(), nomes)))
```

```python
'''6 - Em sistemas de precificação, reajustes são aplicados automaticamente a listas de preços. Leia 8 valores de preços informados pelo usuário e utilize map() com uma lambda para aplicar um acréscimo de 15% a cada item. Exiba os preços originais e os reajustados com duas casas decimais.'''

precos = [36.50,22.40,21.20,40.20,12.40]
print(f' {precos} -> {list(map(lambda x: round(x*1.15,2),precos))}')
#list(map(funcao,lista))
```
