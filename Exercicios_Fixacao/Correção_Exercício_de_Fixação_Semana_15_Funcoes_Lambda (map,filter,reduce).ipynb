```python
'''15 - Em matemática computacional, o produto acumulado (fatorial generalizado) é calculado iterativamente. Utilize reduce() com uma lambda para calcular o produto de todos os elementos de uma lista de 6 inteiros positivos fornecidos pelo usuário. Exiba também qual seria o resultado se um dos valores fosse zero, discutindo o comportamento da função.'''

from functools import reduce

lista = [1, 2, 3, 4, 5, 0]

# Verifica se o 0 está na lista
check = lambda x: True if 0 in lista else False

# Executa a lógica do fatorial/multiplicação
fatorial = lambda x: 'Valor Zero multiplicando sempre dá zero' if check(x) is True else reduce(lambda a, b: a * b, filter(lambda y: y > 0, lista))

# ALTERAÇÃO: Chamando a função passando a lista como argumento
print(fatorial(lista))
```

```python
'''16 - Uma aplicação elegante de reduce() é encontrar extremos de uma coleção sem funções nativas. Leia 9 números do usuário e utilize reduce() com uma lambda que compare dois valores e retorne o maior deles, encontrando assim o valor máximo da lista. Valide o resultado comparando com a função max().
'''
from functools import reduce

# 1. Lendo os 9 números do usuário
numeros = []
print("Digite 9 números:")
for i in range(9):
    num = float(input(f"Número {i+1}: "))
    numeros.append(num)

# 2. Utilizando reduce com lambda para encontrar o valor máximo
# A lambda recebe o acumulador (o maior até agora) e o próximo item, e retorna o maior deles
max_com_reduce = reduce(lambda x, y: x if x > y else y, numeros)

# 3. Validando o resultado com a função nativa max()
max_nativo = max(numeros)

# 4. Exibindo os resultados e a validação
print("\n--- Resultado ---")
print(f"Lista digitada: {numeros}")
print(f"Maior valor encontrado com reduce(): {max_com_reduce}")
print(f"Maior valor encontrado com max(): {max_nativo}")

if max_com_reduce == max_nativo:
    print("✨ Validação bem-sucedida! Os resultados são iguais.")
else:
    print("❌ Algo deu errado na validação.")
```

```python
from functools import reduce

'''
18 - A ordem das operações em um pipeline altera o conjunto de resultados. Leia 8 números do usuário,
utilize map() para elevar cada valor ao quadrado e depois aplique filter() para manter apenas
os resultados maiores que 100. Compare o resultado com a ordem inversa (filter → map) e discuta a diferença.
'''

# Lendo 8 números do usuário
print("Digite 8 números para o Exercício 18:")
numeros = [float(input(f"Número {i+1}: ")) for i in range(8)]

# Ordem 1: Map (Quadrado) -> Filter (> 100)
quadrados = list(map(lambda x: x**2, numeros))
resultado_ordem_1 = list(filter(lambda x: x > 100, quadrados))

# Ordem 2: Filter (> 100) -> Map (Quadrado)
filtrados_antes = list(filter(lambda x: x > 100, numeros))
resultado_ordem_2 = list(map(lambda x: x**2, filtrados_antes))

print("\n--- Resultados Exercício 18 ---")
print(f"Lista original: {numeros}")
print(f"Ordem 1 (Map -> Filter): {resultado_ordem_1}")
print(f"Ordem 2 (Filter -> Map): {resultado_ordem_2}")
```

```python
from functools import reduce

'''
19 — Filter → Map → Reduce (Pipeline Completo)
O encadeamento de filter → map → reduce é o pipeline funcional clássico. Leia 12 números inteiros
do usuário: (1) utilize filter() para manter apenas os valores positivos; (2) utilize map() para
elevar cada positivo ao quadrado; (3) utilize reduce() para obter a soma total dos quadrados.
Exiba cada etapa intermediária do pipeline.
'''

print("\nDigite 12 números inteiros para o Exercício 19:")
numeros_19 = [int(input(f"Inteiro {i+1}: ")) for i in range(12)]

# Etapa 1: Filter (Apenas positivos)
positivos = list(filter(lambda x: x > 0, numeros_19))

# Etapa 2: Map (Elevar ao quadrado)
quadrados_positivos = list(map(lambda x: x**2, positivos))

# Etapa 3: Reduce (Soma total)
soma_total = reduce(lambda x, y: x + y, quadrados_positivos, 0)

print("\n--- Etapas do Pipeline (Exercício 19) ---")
print(f"1. Dados Originais : {numeros_19}")
print(f"2. Filtro (Positivos): {positivos}")
print(f"3. Mapeamento (Quadr.) : {quadrados_positivos}")
print(f"4. Redução (Soma Final): {soma_total}")
```

```python
from functools import reduce

'''
20 - Desafio FINAL — Análise Funcional de Dados Reais
Uma empresa registrou as vendas diárias de 15 dias. Usando exclusivamente lambda, map(), filter()
e reduce() — sem laços for ou while: (a) aplique um desconto de 10% a todos os dias com venda
acima de R$ 500; (b) filtre apenas os dias lucrativos (venda final > R$ 300); (c) calcule o total
acumulado dos dias lucrativos. Leia os 15 valores do usuário e apresente cada etapa do processamento.
'''

print("\nDigite as vendas dos 15 dias:")
# Leitura estritamente funcional de 15 entradas usando list comprehension para o input
vendas_originais = list(map(float, [input(f"Venda Dia {i+1}: R$ ") for i in range(15)]))

# (a) Aplica desconto de 10% se a venda for acima de 500
vendas_com_desconto = list(map(lambda v: v * 0.90 if v > 500 else v, vendas_originais))

# (b) Filtra apenas os dias lucrativos (venda pós-desconto > 300)
dias_lucrativos = list(filter(lambda v: v > 300, vendas_com_desconto))

# (c) Calcula o total acumulado dos dias lucrativos
total_acumulado = reduce(lambda acumulador, valor: acumulador + valor, dias_lucrativos, 0.0)

print("\n================ DESAFIO FINAL ================")
print(f"Vendas Iniciais: {vendas_originais}")
print(f"Etapa A (Desconto Aplicado): {vendas_com_desconto}")
print(f"Etapa B (Dias Lucrativos > 300): {dias_lucrativos}")
print(f"Etapa C (Faturamento Acumulado): R$ {total_acumulado:.2f}")
print("===============================================")
```
