```python
#01 - Crie uma lista contendo os nomes de 5 vilões clássicos do Mario.
#Utilize a função len() para contar o número total de itens e exiba o resultado no terminal.
viloes = ['Bowser','Bowser Jr.','Kamek','Donkey Kong','Iggy Koopa']
contador = len(viloes)
print(f'A quantidade de itens é {contador}')
```

```python
#02 - Inicie uma lista vazia chamada moedas_coletadas.
#Use um laço de repetição para pedir que o usuário informe 4 valores inteiros
#e adicione cada um deles ao final da lista usando append().
moedas_coletadas = []
for i in range(4):
    valor = int(input(f'Adicione o {i+1}° valor:'))
    moedas_coletadas.append(valor)
print(moedas_coletadas)
```

```python
#03 - Crie a lista herois = ['Mario', 'Peach'].
#Utilize o método insert() para adicionar o nome 'Luigi' exatamente no meio da lista (índice 1)
#e exiba a lista final.
herois = ['Mario','Peach']
herois.insert(len(herois)//2,'Luigi')
print(herois)
```

```python
#04 - Dada a lista viloes = ['Bowser', 'Goomba', 'Wario', 'Koopa'],
#solicite que o usuário digite um nome.
#Se o nome estiver na lista, use remove() para excluí-lo.
viloes = ['Bowser', 'Goomba', 'Wario', 'Koopa']
nome = input('Digite um nome de vilão do Mário: ').capitalize()
if nome in viloes:
    viloes.remove(nome)
    print(viloes)
else:
    print(viloes)
```

```python
#05 - Crie uma lista com nomes de 4 itens de poder (ex: Estrela, Cogumelo).
#Use pop() para remover o último item e mostre qual item foi removido
#e como a lista ficou.

powers = ['Estrela','Cogumelo Vermelho','Flor de Fogo','Pena']
removido = powers[-1] #Armazena o valor a ser removido (último)
powers.pop() #Remove o último elemento
print(f'Foi removido {removido} \ne a lista ficou: {powers}!')
```

```python
#06 - Peça 5 pontuações ao usuário e armazene-as.
#Utilize sort() para ordenar a lista em ordem crescente
#e exiba-a organizada.

pontuacoes =[] #Cria a lista vazia
for i in range(5): #Repete 5 vezes para anotar as pontuações
    #Capta a nota
    ponto = float(input(f'Digite a {i+1}ª nota:'))
    #adiciona a nota na lista
    pontuacoes.append(ponto)
#Após todas as notas na lista, ordena a lista
pontuacoes.sort()
#Imprime
print(pontuacoes)
```

```python
#07 - Crie uma lista com 4 nomes de fases.
#Use enumerate() dentro de um for para exibir o índice e o nome de cada fase
#(Ex: "Fase 0: Castelo de Fogo").
fases = ['Castelo de Fogo Fogoso','Ilha das Nuvens Nuvemnosas','Galáxia Galaxiosa','Deserto Desertoso']
for num, fase in enumerate (fases):
    print(f'Fase {num} - {fase}')
```

```python
#08 - Crie uma lista vazia.
#Peça nomes de poderes ao usuário até que ele digite "sair".
#Use append() para adicionar e,
#no fim, use len() para mostrar quantos poderes foram coletados.

poderes = []
while True:
    poder = input('Digite o novo poder: ')
    if poder.lower() == 'sair':
        print(poderes)
        break
    else:
        poderes.append(poder.capitalize())
        print(f'Poder - {poder} foi adicionado a lista!')
```

```python
#09 - Crie uma lista com 5 nomes de personagens.
#Peça ao usuário um número de 0 a 4 e use pop(indice) para remover o personagem daquela posição.
personagens = ['Mario','Luigi','Peach','Toad','Yoshi']
indice = int(input('Digite um valor de 0 a 4:'))
if 0 <= indice <= 4:
    personagens.pop(indice) #Remove pelo índice
    print(personagens)
else:
    print('Valor inválido')
```
