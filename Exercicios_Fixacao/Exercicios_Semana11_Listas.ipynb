```python
#1 - Crie uma lista chamada notas com 5 notas de 0 a 10. Depois:
#(a) imprima a primeira e a última nota;
#(b) altere a segunda nota para 10;
#(c) adicione uma nova nota com append();
#(d) imprima o total de notas com len().

notas = [6,6.5,7,5,9]
print(f'Lista base: {notas}')

print(f'a)Primeira:{notas[0]} e Última: {notas[-1]}')

notas[1] = 10
print(f'b){notas}')

notas.append(8)
print(f'c){notas}')

print(f'd)Total de notas: {len(notas)}')

#Plus Time:
#Ordem Crescente
notas.sort()
print(f'Crescente: {notas}')

#Ordem Descrescente
notas.sort(reverse=True)
print(f'Decrescente: {notas}')
```

```python
#2 - Crie uma lista chamada turma com 4 nomes de colegas.
#Use um for para imprimir: 'Presente: [nome]' para cada aluno.
#Depois, adicione seu próprio nome com append() e refaça a chamada.

turma = ['Aragorn','Legolas','Gimlin','Faramir']
#a)
print('Letra A:')
for aluno in turma:
    print(f'Presente: {aluno}')
turma.append('Glorioso Mestre Gustavo, Rei de Valfenda')
#b)
print('\nLetra B:')
for aluno in turma:
    print(f'Presente: {aluno}')
```

```python
#3 - Crie uma lista chamada valores com pelo menos 6 números inteiros;
# em seguida, remova um elemento específico usando remove(),
# remova o último elemento com pop(),
# ordene a lista com sort() e, por fim,
# utilize um for para exibir cada valor da lista junto com sua posição (índice).

valores = [8,7,6,5,4]
print(f'a) {valores}')

valores = [8,7,6,5,4]
valores.remove(6)
print(f'b) {valores}')

valores = [8,7,6,5,4]
valores.pop()
print(f'c) {valores}')

valores = [8,7,6,5,4]
valores.sort()
print(f'd) {valores}')

print('e)')
for i, valor in enumerate (valores):
    print(f'   {i+1}° número: {valor}')

```
