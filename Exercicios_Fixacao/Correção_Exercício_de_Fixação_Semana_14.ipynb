```python
#3 - Contabilizar a ocorrência de elementos em uma sequência é uma
#tarefa recorrente em análise de dados e processamento de texto.
#Escreva um programa que receba uma lista de palavras (definida no
#código) e construa um dict que mapeie cada palavra à sua frequência de
#aparição na lista. Após construir o dicionário, converta-o em uma lista
#de tuplas (palavra, frequência), ordene-a pela frequência em ordem
#decrescente e exiba os cinco elementos mais frequentes. Por fim,
#extraia apenas as palavras (chaves) como um conjunto (set) e exiba o
#vocabulário único identificado.

palavras = ['Abacate','Banana','Kiwi','Maçã','Uva','Abacate','Banana','Maçã','Abacate','Maçã','Maçã','Melancia','Uva']
dic_palavras = {}
for palavra in palavras:
    dic_palavras[palavra] = palavras.count(palavra)
print(f'A - Dicionário de palavra e frequencia:\n{dic_palavras}')
tuplas = tuple(sorted(dic_palavras.items(), key=lambda item : item[1], reverse=True))
print(f'\nB - Tuplas em ordem descrescente:\n{tuplas}')
print('\nC - 5 itens que mais se repetem:')
for i in range(5):
    print(tuplas[i])
print('\nD - Conjunto das chaves:')
final = set(dict(tuplas).keys())
print(final)
```

```python
'''5 - Operações de teoria dos conjuntos são amplamente aplicadas para
comparar e combinar coleções de dados. Defina dois dicionários que
representem o estoque de dois armazéns distintos, onde as chaves são
nomes de produtos e os valores são quantidades disponíveis. Converta as
chaves de cada dicionário em conjuntos (set) e utilize as operações de
união, interseção e diferença para identificar: (a) todos os produtos
disponíveis em pelo menos um armazém; (b) os produtos presentes em ambos;
(c) os produtos exclusivos de cada armazém. Exiba os resultados de forma
clara e estruturada.'''
# 1. Definição dos dicionários que representam o estoque dos dois armazéns
armazem_A = {
    "Notebook": 15,
    "Mouse": 50,
    "Teclado": 30,
    "Monitor": 12,
    "Headset": 25
}

armazem_B = {
    "Monitor": 8,
    "Headset": 15,
    "Webcam": 20,
    "Suporte Articulado": 10,
    "Mouse": 40
}

# 2. Conversão das chaves dos dicionários em conjuntos (sets)
produtos_A = set(armazem_A.keys())
produtos_B = set(armazem_B.keys())

# 3. Operações de teoria dos conjuntos
# (a) União: todos os produtos disponíveis em pelo menos um armazém
todos_produtos = produtos_A.union(produtos_B)

# (b) Interseção: produtos presentes em ambos os armazéns
produtos_em_ambos = produtos_A.intersection(produtos_B)

# (c) Diferença: produtos exclusivos de cada armazém
exclusivos_A = produtos_A.difference(produtos_B)
exclusivos_B = produtos_B.difference(produtos_A)

# 4. Exibição dos resultados de forma clara e estruturada
print("RELATÓRIO DE ESTOQUE COMPATIVEL (TEORIA DOS CONJUNTOS)")

print(f"\nProdutos no Armazém A: {produtos_A}")
print(f"Produtos no Armazém B: {produtos_B}")

print("\n(a) Todos os produtos disponíveis em pelo menos um armazém (União):")
print(todos_produtos)

print("\n(b) Produtos presentes em ambos os armazéns (Interseção):")
print(produtos_em_ambos)

print("\n(c) Produtos exclusivos de cada armazém (Diferença):")
print(f" -> Exclusivos do Armazém A: {exclusivos_A}")
print(f" -> Exclusivos do Armazém B: {exclusivos_B}")
```

```python
'''6 - Inverter o mapeamento de um dicionário — tornando os valores em
chaves e as chaves em valores — é uma operação útil em sistemas de
indexação reversa. Escreva um programa que parta de um dicionário
cujas chaves são códigos numéricos e os valores são nomes de categorias.
Inverta esse dicionário de modo que os nomes se tornem chaves e
os códigos se tornem valores. Em seguida, converta o dicionário
invertido em uma lista de tuplas ordenadas alfabeticamente pelas chaves
e exiba o resultado. Trate o caso em que dois códigos diferentes possam
ter o mesmo nome, agrupando os códigos em uma lista associada a cada nome.
'''
# 1. Definição do dicionário inicial (com códigos e categorias)
# Note que "Eletrônicos" e "Móveis" aparecem duplicados com códigos diferentes
categorias_originais = {
    101: "Eletrônicos",
    102: "Vestuário",
    103: "Móveis",
    104: "Eletrônicos",
    105: "Livros",
    106: "Móveis"
}

# 2. Dicionário vazio para armazenar o mapeamento invertido
dicionario_invertido = {}

# 3. Processo de inversão agrupando códigos repetidos em listas
for codigo, nome in categorias_originais.items():
    # Se o nome já existe no novo dicionário, adicionamos o código à lista existente
    if nome in dicionario_invertido:
        dicionario_invertido[nome].append(codigo)
    # Se o nome ainda não existe, criamos uma nova lista com o código atual
    else:
        dicionario_invertido[nome] = [codigo]

# 4. Conversão para uma lista de tuplas e ordenação alfabética pelas chaves (nomes)
# O método sorted() em dicionario_invertido.items() ordena automaticamente pelas chaves
lista_ordenada = sorted(dicionario_invertido.items())

# 5. Exibição dos resultados de forma clara e estruturada
print("INDEXAÇÃO REVERSA DE CATEGORIAS")
print("\nDicionário Original (Código -> Categoria):")
print(categorias_originais)

print("\nDicionário Invertido e Agrupado (Categoria -> Lista de Códigos):")
print(dicionario_invertido)

print("\nLista de Tuplas Ordenada Alfabeticamente:")
for categoria, codigos in lista_ordenada:
    print(f"Categoria: {categoria:<15} | Códigos: {codigos}")
```

```python
'''7 - Agrupar elementos de uma coleção segundo um critério derivado
dos próprios dados é uma operação central em análise e transformação
de dados. Dado uma lista de tuplas no formato (produto, categoria, preço)
com ao menos 10 entradas, escreva um programa que construa um dicionário
onde cada chave é uma categoria e o valor associado é a lista de tuplas
pertencentes àquela categoria. Em seguida, para cada categoria, calcule
o preço médio dos produtos e exiba um resumo no formato categoria: N
produtos, preço médio R$ X. Utilize exclusivamente operações de
conversão entre as estruturas list, dict e tuple.
'''

# 1. Lista de tuplas inicial (mínimo de 10 entradas) no formato (produto, categoria, preço)
dados_produtos = [
    ("Notebook", "Eletrônicos", 4500.00),
    ("Mouse", "Eletrônicos", 150.00),
    ("Camisa", "Vestuário", 89.90),
    ("Teclado", "Eletrônicos", 250.00),
    ("Sofá", "Móveis", 1800.00),
    ("Calça", "Vestuário", 120.00),
    ("Monitor", "Eletrônicos", 1200.00),
    ("Mesa", "Móveis", 650.00),
    ("Tênis", "Vestuário", 299.90),
    ("Cadeira", "Móveis", 450.00)
]

# 2. Construção do dicionário agrupado por categoria
categorias_agrupadas = {}

for produto, categoria, preco in dados_produtos:
    # Se a categoria já existe, adiciona a tupla completa à lista daquela categoria
    if categoria in categorias_agrupadas:
        categorias_agrupadas[categoria].append((produto, categoria, preco))
    # Se não existe, cria a lista com a primeira tupla
    else:
        categorias_agrupadas[categoria] = [(produto, categoria, preco)]

# 3. Cálculo da média e exibição do resumo formatado
print("RESUMO DE PRODUTOS POR CATEGORIA E PREÇO MÉDIO")
for categoria, lista_tuplas in categorias_agrupadas.items():
    # Contador de produtos na categoria atual
    total_produtos = len(lista_tuplas)

    # Soma dos preços dos produtos da categoria atual
    soma_precos = 0
    for produto, cat, preco in lista_tuplas:
        soma_precos += preco

    # Cálculo da média aritmética simples
    preco_medio = soma_precos / total_produtos

    # Exibição do resumo conforme solicitado no enunciado
    print(f"{categoria}: {total_produtos} produtos, preço médio R$ {preco_medio:.2f}")

```

```python
''' 9 - Estruturas aninhadas são comuns em respostas de APIs e arquivos
de configuração. Dado um dicionário cujas chaves são nomes de alunos e
os valores são dicionários internos no formato {"notas": [lista de notas]
, "turma": "X"}, escreva um programa que realize as seguintes
transformações: (a) converta cada lista de notas em uma tupla imutável;
(b) calcule a média de cada aluno e adicione-a ao dicionário interno;
(c) extraia todos os nomes de alunos com média superior a 7,0 como um
set; (d) gere uma lista de tuplas (nome, média, turma) ordenada pela
média em ordem decrescente e exiba o ranking final.
'''
# 1. Definição das três fontes de dados com sobreposições e um conflito intencional no ID 3
fonte_1 = [(1, "Ana"), (2, "Bruno"), (3, "Carlos")]
fonte_2 = [(2, "Bruno"), (3, "Carlos"), (4, "Daniela")]
fonte_3 = [(3, "Carla"), (5, "Eduardo"), (1, "Ana")]

# (a) Unir todas as listas em uma única
lista_unida = fonte_1 + fonte_2 + fonte_3

# (b) Converter em um conjunto de tuplas para eliminar duplicatas exatas
conjunto_tuplas = set(lista_unida)

# (c) Converter em dicionário tratando conflitos (mantendo o primeiro e registrando o conflito)
dicionario_final = {}
lista_conflitos = []

for id_num, nome in conjunto_tuplas:
    if id_num in dicionario_final:
        if dicionario_final[id_num] != nome:
            lista_conflitos.append(f"Conflito no ID {id_num}: Mantido '{dicionario_final[id_num]}', rejeitado '{nome}'")
    else:
        dicionario_final[id_num] = nome

# (d) Ordenar o dicionário final pela chave numérica
dicionario_ordenado = sorted(dicionario_final.items())

# Exibição dos resultados de forma clara e estruturada
print("INTEGRAÇÃO E FUSÃO DE FONTES DE DADOS")

print(f"\n(a) Lista Unida (Total de {len(lista_unida)} registros):")
print(lista_unida)

print(f"\n(b) Conjunto (Sem duplicatas exatas - Restaram {len(conjunto_tuplas)}):")
print(conjunto_tuplas)

print("\n(c) Registro de Conflitos Encontrados:")
if lista_conflitos:
    for conflito in lista_conflitos:
        print(f" [AVISO] {conflito}")
else:
    print(" Nenhum conflito de nome encontrado.")

print("\n(d) Dicionário Final Integrado e Ordenado por ID:")
for id_num, nome in dicionario_ordenado:
    print(f"ID: {id_num:<4} | Nome: {nome}")
```

```python
'''10 - Em engenharia de dados, pipelines encadeiam múltiplas
transformações sobre uma mesma coleção. Parta de uma lista de strings
no formato "nome:valor" com ao menos 12 entradas e valores possivelmente
repetidos. Implemente as seguintes etapas em sequência: (a) converta a
lista em um conjunto de tuplas (nome, valor) eliminando entradas
duplicadas; (b) a partir do conjunto, construa um dicionário que
mapeie cada nome à soma acumulada de todos os seus valores;
(c) converta o dicionário em uma lista de tuplas (nome, soma) e filtre
apenas aqueles com soma superior à média geral; (d) armazene o resultado
final como uma tupla imutável e exiba cada etapa intermediária com seu
respectivo tipo (type()) e conteúdo.
'''

# Dados iniciais: Lista de strings no formato "nome:valor" (12 entradas com repetições)
dados_brutos = [
    "SensorA:10", "SensorB:20", "SensorA:10",
    "SensorC:30", "SensorB:15", "SensorA:25",
    "SensorD:5",  "SensorC:30", "SensorD:15",
    "SensorE:50", "SensorE:10", "SensorF:12"
]

print("ETAPA INICIAL")
print(f"Tipo: {type(dados_brutos)}")
print(f"Conteúdo: {dados_brutos}\n")


# --- (a) Converter em conjunto de tuplas (nome, valor) eliminando duplicadas ---
conjunto_tuplas = set()
for item in dados_brutos:
    nome, valor_str = item.split(":")
    conjunto_tuplas.add((nome, int(valor_str)))

print("ETAPA (a) - Conjunto de Tuplas (Sem Duplicatas)")
print(f"Tipo: {type(conjunto_tuplas)}")
print(f"Conteúdo: {conjunto_tuplas}\n")


# --- (b) Construir dicionário com a soma acumulada por nome ---
dicionario_somas = {}
for nome, valor in conjunto_tuplas:
    if nome in dicionario_somas:
        dicionario_somas[nome] += valor
    else:
        dicionario_somas[nome] = valor

print("ETAPA (b) - Dicionário de Somas Acumuladas")
print(f"Tipo: {type(dicionario_somas)}")
print(f"Conteúdo: {dicionario_somas}\n")


# --- (c) Converter para lista de tuplas e filtrar acima da média geral ---
# Passo 1: Calcular a média geral das somas acumuladas
total_somas = 0
quantidade_nomes = len(dicionario_somas)

for soma in dicionario_somas.values():
    total_somas += soma

media_geral = total_somas / quantidade_nomes

# Passo 2: Filtrar os itens que estão acima da média calculada
lista_filtrada = []
for nome, soma in dicionario_somas.items():
    if soma > media_geral:
        lista_filtrada.append((nome, soma))

print("ETAPA (c) - Lista Filtrada (Apenas valores acima da média)")
print(f"Média Geral calculada: {media_geral}")
print(f"Tipo: {type(lista_filtrada)}")
print(f"Conteúdo: {lista_filtrada}\n")


# --- (d) Armazenar o resultado final como uma tupla imutável ---
resultado_final = tuple(lista_filtrada)

print("ETAPA (d) - Resultado Final (Estrutura Imutável)")
print(f"Tipo: {type(resultado_final)}")
print(f"Conteúdo: {resultado_final}")
```
