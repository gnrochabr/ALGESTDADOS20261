```python
inscritos = ['Ana', 'Bruno', 'Ana', 'Carla', 'Bruno', 'Diego']

# Remove duplicatas usando set()
inscritos_unicos = set(inscritos)

# Exibe os inscritos únicos
print(inscritos_unicos)

# Quantidade de inscritos únicos
print(len(inscritos_unicos))
```

```python
turma_python = {'Ana', 'Bruno', 'Carla', 'Diego'}
turma_java = {'Carla', 'Diego', 'Elena', 'Fábio'}

# a) Interseção
print("Nas duas turmas:")
print(turma_python.intersection(turma_java))

# b) União
print("\nEm pelo menos uma turma:")
print(turma_python.union(turma_java))

# c) Diferença
print("\nSomente em Python:")
print(turma_python.difference(turma_java))
```

```python
notas = [7, 8, 7, 9, 10, 8, 6, 10]

# Remove duplicatas
notas_unicas = set(notas)

# Ordena e converte para lista
lista_ordenada = sorted(notas_unicas)

# Exibe a lista final
print(lista_ordenada)

# Quantidade de notas únicas
print(len(lista_ordenada))
```
