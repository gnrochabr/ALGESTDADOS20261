```python
#8 - Dada a string "o gato viu o rato e o rato correu do gato", escreva um programa que conte
#quantas vezes cada palavra aparece e armazene o resultado em um dicionário.
#Trate tudo em letras minúsculas. Exiba os pares palavra–frequência ao final.
#Use .split() para separar as palavras.


frase = "o gato viu o rato e o rato correu do gato" #frase é uma lista de caracteres
lista_palavras = frase.split(' ') #cria a lista de palavras
conjunto_palavras = sorted(set(lista_palavras)) #lista ordenada em conjunto
dicionario_palavras = {} #cria o dicionário vazio

for palavra in conjunto_palavras:
    contador = 0
    for ocorrencia in lista_palavras:
        if palavra == ocorrencia:
            contador+=1
    dicionario_palavras[palavra] = contador

for palavra, quantidade in dicionario_palavras.items():
    print(f'{palavra}: {quantidade}')
```

```python
#11 - Dado o dicionário precos = {"caneta": 3.50, "mochila": 129.90, "borracha": 2.00, "fone": 89.99, "caderno": 18.00, "tablet": 749.00},
#crie um novo dicionário contendo apenas os produtos com preço superior a R$ 50,00 usando compreensão de dicionário.
#Calcule e exiba também o preço médio dos itens filtrados.# Dicionário original
precos = {
    "caneta": 3.50,
    "mochila": 129.90,
    "borracha": 2.00,
    "fone": 89.99,
    "caderno": 18.00,
    "tablet": 749.00
}

# Criando dicionário vazio
produtos_filtrados = {}

# Variável para armazenar a soma dos valores
soma = 0

# Filtrando produtos com preço maior que R$ 50,00
for chave, valor in precos.items():
    if valor > 50:
        produtos_filtrados[chave] = valor
        soma += valor

# Calculando a média dos preços filtrados
media = soma / len(produtos_filtrados)

# Exibindo resultados
print("Produtos filtrados:")
print(produtos_filtrados)

print(f"\nSoma dos valores: R$ {soma:.2f}")
print(f"Preço médio dos itens filtrados: R$ {media:.2f}")
```

```python
#12 - Dados dois dicionários de estoque: filial_a = {"camiseta": 30, "calca": 15, "tenis": 10} e
#filial_b = {"tenis": 5, "bone": 20, "camiseta": 12}, mescle-os em um único dicionário.
#Quando um produto aparecer nas duas filiais, some as quantidades. Exiba o estoque consolidado.

# Dicionários das filiais
filial_a = {
    "camiseta": 30,
    "calca": 15,
    "tenis": 10
}

filial_b = {
    "tenis": 5,
    "bone": 20,
    "camiseta": 12
}

# Dicionário consolidado
estoque = {}

# Percorrendo a filial A
for chave, valor in filial_a.items():

    # Verifica se o produto também existe na filial B
    if chave in filial_b:
        estoque[chave] = valor + filial_b[chave]

    # Caso não exista, mantém o valor da filial A
    else:
        estoque[chave] = valor

# Percorrendo a filial B para adicionar produtos
# que não existem na filial A
for chave, valor in filial_b.items():

    if chave not in filial_a:
        estoque[chave] = valor

# Exibindo resultado
print("Estoque consolidado:")
print(estoque)
```

```python
#13 - Crie um sistema interativo de agenda no terminal com as opções: (1) Cadastrar contato (nome → telefone);
#(2) Buscar contato pelo nome; (3) Remover contato; (4) Listar todos os contatos; (5) Sair.
#Use um dicionário para armazenar os dados e um laço while para manter o menu ativo.
# Dicionário da agenda
agenda = {}

# Menu principal
while True:

    print("\n=== AGENDA DE CONTATOS ===")
    print("1 - Cadastrar contato")
    print("2 - Buscar contato")
    print("3 - Remover contato")
    print("4 - Listar contatos")
    print("5 - Sair")

    opcao = int(input("Escolha uma opção: "))

    match opcao:

        # Cadastrar contato
        case 1:

            nome = input("Digite o nome: ").lower().capitalize()
            telefone = input("Digite o telefone: ")

            agenda[nome] = telefone

            print("Contato cadastrado com sucesso!")

        # Buscar contato
        case 2:

            nome = input("Digite o nome do contato: ").lower().capitalize()

            if nome in agenda:
                print(f"Telefone: {agenda[nome]}")
            else:
                print("Contato não encontrado!")

        # Remover contato
        case 3:

            nome = input("Digite o nome do contato: ").lower().capitalize()

            if nome in agenda:
                del agenda[nome]
                print("Contato removido!")
            else:
                print("Contato não encontrado!")

        # Listar contatos
        case 4:

            print("\n=== CONTATOS ===")

            if len(agenda) == 0:
                print("Agenda vazia!")

            else:
                for chave, valor in agenda.items():
                    print(f"{chave} -> {valor}")

        # Sair
        case 5:

            print("Encerrando agenda...")
            break

        # Opção inválida
        case _:

            print("Opção inválida!")
```

```python
#14 - Dada a lista linhas = ["Ana,9.5", "Bruno,6.0", "Carla,8.3", "Diego,7.1", "Eva,10.0"], que simula linhas de um arquivo CSV,
#construa um dicionário onde cada nome é chave e a nota (convertida para float) é o valor. Exiba o nome do aluno com a maior nota e o com a menor nota.
# Lista simulando linhas de um CSV
linhas = [
    "Ana,9.5",
    "Bruno,6.0",
    "Carla,8.3",
    "Diego,7.1",
    "Eva,10.0"
]

# Dicionário de alunos e notas
notas = {}

# Variáveis de controle
maior_nome = ""
menor_nome = ""

maior_nota = 0
menor_nota = 10

# Percorrendo as linhas
for linha in linhas:

    # Separando nome e nota
    dados = linha.split(",")

    nome = dados[0]
    nota = float(dados[1])

    # Adicionando ao dicionário
    notas[nome] = nota

    # Verificando maior nota
    if nota > maior_nota:
        maior_nota = nota
        maior_nome = nome

    # Verificando menor nota
    if nota < menor_nota:
        menor_nota = nota
        menor_nome = nome

# Exibindo resultados
print("Dicionário de notas:")
print(notas)

print(f"\nMaior nota: {maior_nome} -> {maior_nota}")
print(f"Menor nota: {menor_nome} -> {menor_nota}")
```

```python
#15 - Crie um dicionário turma onde cada chave é o nome de um aluno e o valor é outro dicionário com notas (lista com 3 valores) e frequencia (porcentagem).
#Percorra a estrutura e exiba, para cada aluno, a média das notas, a frequência e se está aprovado (média ≥ 6.0 e frequência ≥ 75%) ou reprovado.
# Dicionário da turma
turma = {
    "Ana": {
        "notas": [8.0, 7.5, 9.0],
        "frequencia": 90
    },

    "Bruno": {
        "notas": [5.0, 6.0, 4.5],
        "frequencia": 80
    },

    "Carla": {
        "notas": [7.0, 6.5, 8.0],
        "frequencia": 70
    },

    "Diego": {
        "notas": [9.5, 8.5, 10.0],
        "frequencia": 95
    }
}

# Percorrendo os alunos
for aluno, dados in turma.items():

    notas = dados["notas"]
    frequencia = dados["frequencia"]

    # Calculando média
    soma = 0

    for nota in notas:
        soma += nota

    media = soma / len(notas)

    # Verificando situação
    if media >= 6.0 and frequencia >= 75:
        situacao = "Aprovado"

    else:
        situacao = "Reprovado"

    # Exibindo resultados
    print(f"Aluno: {aluno}")
    print(f"Média: {media:.2f}")
    print(f"Frequência: {frequencia}%")
    print(f"Situação: {situacao}\n")

```

```python
#16 - Dada uma lista de tuplas (categoria, valor) representando transações financeiras (ex.: "alimentação", "transporte", "lazer"),
#construa um dicionário onde cada chave é uma categoria e o valor é uma lista com todos os gastos daquela categoria.
#Exiba o total gasto por categoria e identifique qual teve o maior gasto acumulado. Use .setdefault().
# Lista de transações
transacoes = [
    ("alimentacao", 45.0),
    ("transporte", 20.0),
    ("lazer", 80.0),
    ("alimentacao", 35.0),
    ("transporte", 15.0),
    ("lazer", 50.0),
    ("alimentacao", 25.0)
]

# Dicionário de categorias
gastos = {}

# Convertendo a lista para dicionário
for categoria, valor in transacoes:

    # Verifica se a categoria já existe
    if categoria not in gastos:
        gastos[categoria] = []

    # Adiciona o valor na lista da categoria
    gastos[categoria].append(valor)

# Variáveis de controle
maior_categoria = ""
maior_total = 0

# Exibindo resultados
for categoria, valores in gastos.items():

    soma = 0

    for valor in valores:
        soma += valor

    print(f"Categoria: {categoria}")
    print(f"Gastos: {valores}")
    print(f"Total gasto: R$ {soma:.2f}")
    print()

    # Verificando maior gasto acumulado
    if soma > maior_total:
        maior_total = soma
        maior_categoria = categoria

# Resultado final
print(f"Maior gasto acumulado: {maior_categoria}")
print(f"Total: R$ {maior_total:.2f}")
```

```python
#17 - Dado um dicionário onde cada chave é o título de um documento e o valor é seu conteúdo textual,
#construa um índice invertido: um novo dicionário onde cada palavra (em minúsculas, sem duplicatas por documento) mapeia para a lista de títulos em que aparece.
#Receba uma palavra do usuário e exiba em quais documentos ela foi encontrada.
# Dicionário de documentos
documentos = {
    "doc1": "Python é uma linguagem poderosa e simples",
    "doc2": "Java e Python são linguagens populares",
    "doc3": "Banco de dados e linguagem SQL"
}

# Índice invertido
indice = {}

# Percorrendo os documentos
for titulo, conteudo in documentos.items():

    # Separando palavras e removendo duplicatas
    palavras = set(conteudo.lower().split())

    # Construindo índice invertido
    for palavra in palavras:

        if palavra not in indice:
            indice[palavra] = []

        indice[palavra].append(titulo)

# Exibindo índice invertido
print("Índice invertido:")
print(indice)

# Busca de palavra
busca = input("\nDigite uma palavra: ").lower()

# Verificando se a palavra existe
if busca in indice:

    print(f"\nA palavra '{busca}' foi encontrada em:")

    for documento in indice[busca]:
        print(documento)

else:
    print("Palavra não encontrada!")
```

```python
#18 - Implemente um sistema de carrinho de compras usando dois dicionários: catalogo (produto → preço) e carrinho (produto → quantidade).
#Permita ao usuário: adicionar itens, remover itens, alterar a quantidade e calcular o total. Ao finalizar, exiba um recibo com os itens, quantidades, subtotais e total geral.
# Catálogo de produtos
catalogo = {
    "Arroz": 25.00,
    "Feijao": 9.50,
    "Macarrao": 6.00,
    "Leite": 5.50,
    "Cafe": 18.00
}

# Carrinho de compras
carrinho = {}

# Menu principal
while True:

    print("\n=== CARRINHO DE COMPRAS ===")
    print("1 - Adicionar item")
    print("2 - Remover item")
    print("3 - Alterar quantidade")
    print("4 - Ver carrinho")
    print("5 - Finalizar compra")

    opcao = int(input("Escolha uma opção: "))

    match opcao:

        # Adicionar item
        case 1:

            print("\n=== CATÁLOGO ===")

            for produto, preco in catalogo.items():
                print(f"{produto} -> R$ {preco:.2f}")

            produto = input("\nDigite o produto: ").capitalize()

            if produto in catalogo:

                quantidade = int(input("Digite a quantidade: "))

                if produto in carrinho:
                    carrinho[produto] += quantidade

                else:
                    carrinho[produto] = quantidade

                print("Produto adicionado!")

            else:
                print("Produto não encontrado!")

        # Remover item
        case 2:

            produto = input("Digite o produto para remover: ").capitalize()

            if produto in carrinho:

                del carrinho[produto]

                print("Produto removido!")

            else:
                print("Produto não está no carrinho!")

        # Alterar quantidade
        case 3:

            produto = input("Digite o produto: ").capitalize()

            if produto in carrinho:

                quantidade = int(input("Nova quantidade: "))

                carrinho[produto] = quantidade

                print("Quantidade atualizada!")

            else:
                print("Produto não está no carrinho!")

        # Ver carrinho
        case 4:

            print("\n=== CARRINHO ===")

            total = 0

            if len(carrinho) == 0:
                print("Carrinho vazio!")

            else:

                for produto, quantidade in carrinho.items():

                    preco = catalogo[produto]
                    subtotal = preco * quantidade

                    total += subtotal

                    print(f"{produto}")
                    print(f"Quantidade: {quantidade}")
                    print(f"Subtotal: R$ {subtotal:.2f}")
                    print()

                print(f"Total parcial: R$ {total:.2f}")

        # Finalizar compra
        case 5:

            print("\n=== RECIBO ===")

            total_geral = 0

            for produto, quantidade in carrinho.items():

                preco = catalogo[produto]
                subtotal = preco * quantidade

                total_geral += subtotal

                print(f"{produto}")
                print(f"Quantidade: {quantidade}")
                print(f"Preço unitário: R$ {preco:.2f}")
                print(f"Subtotal: R$ {subtotal:.2f}")
                print()

            print(f"TOTAL GERAL: R$ {total_geral:.2f}")
            print("\nCompra finalizada!")

            break

        # Opção inválida
        case _:

            print("Opção inválida!")
```

```python
#19 - Leia uma string digitada pelo usuário e gere um histograma de frequência de cada caractere (ignorando espaços).
#Exiba os resultados no terminal de forma visual, onde cada caractere é seguido por uma barra de asteriscos proporcional à sua frequência.
#Ordene do mais frequente para o menos frequente usando sorted() com key.
# Lendo a string do usuário
texto = input("Digite uma frase: ").lower()

# Dicionário de frequência
frequencia = {}

# Contando caracteres
for caractere in texto:

    # Ignorando espaços
    if caractere != " ":

        if caractere not in frequencia:
            frequencia[caractere] = 0

        frequencia[caractere] += 1

# Função para retornar a frequência
def obter_frequencia(item):
    return item[1]

# Ordenando do mais frequente para o menos frequente
ordenado = sorted(
    frequencia.items(),
    key=obter_frequencia,
    reverse=True
)

# Exibindo histograma
print("\n=== HISTOGRAMA ===")

for caractere, quantidade in ordenado:

    barra = "*" * quantidade

    print(f"{caractere} -> {barra} ({quantidade})")
```

```python
#20 - Construa um mini banco de dados de clientes com um dicionário aninhado (ID → dados do cliente).
#Implemente as funções cadastrar_cliente(), depositar(), sacar() (com validação de saldo), extrato() (exibe nome, saldo e histórico) e maior_saldo() — tudo sem o uso de classes.
#Demonstre com pelo menos 3 clientes e diversas transações.

# Banco de dados de clientes
clientes = {}

# =========================
# CADASTRANDO CLIENTES
# =========================

clientes[1] = {
    "nome": "Ana",
    "saldo": 0,
    "historico": []
}

clientes[2] = {
    "nome": "Bruno",
    "saldo": 0,
    "historico": []
}

clientes[3] = {
    "nome": "Carla",
    "saldo": 0,
    "historico": []
}

# =========================
# DEPÓSITOS
# =========================

clientes[1]["saldo"] += 500
clientes[1]["historico"].append("Depósito de R$ 500.00")

clientes[2]["saldo"] += 300
clientes[2]["historico"].append("Depósito de R$ 300.00")

clientes[3]["saldo"] += 1000
clientes[3]["historico"].append("Depósito de R$ 1000.00")

# =========================
# SAQUES
# =========================

valor = 120

if valor <= clientes[1]["saldo"]:

    clientes[1]["saldo"] -= valor

    clientes[1]["historico"].append(
        f"Saque de R$ {valor:.2f}"
    )

else:
    print("Saldo insuficiente!")

valor = 50

if valor <= clientes[2]["saldo"]:

    clientes[2]["saldo"] -= valor

    clientes[2]["historico"].append(
        f"Saque de R$ {valor:.2f}"
    )

else:
    print("Saldo insuficiente!")

valor = 200

if valor <= clientes[3]["saldo"]:

    clientes[3]["saldo"] -= valor

    clientes[3]["historico"].append(
        f"Saque de R$ {valor:.2f}"
    )

else:
    print("Saldo insuficiente!")

# =========================
# NOVOS DEPÓSITOS
# =========================

clientes[1]["saldo"] += 250
clientes[1]["historico"].append("Depósito de R$ 250.00")

clientes[2]["saldo"] += 100
clientes[2]["historico"].append("Depósito de R$ 100.00")

# =========================
# EXTRATOS
# =========================

for id_cliente, dados in clientes.items():

    print("\n=== EXTRATO ===")

    print(f"ID: {id_cliente}")
    print(f"Nome: {dados['nome']}")
    print(f"Saldo: R$ {dados['saldo']:.2f}")

    print("\nHistórico:")

    for item in dados["historico"]:
        print(item)

# =========================
# MAIOR SALDO
# =========================

maior_cliente = ""
maior_saldo = 0

for id_cliente, dados in clientes.items():

    if dados["saldo"] > maior_saldo:

        maior_saldo = dados["saldo"]
        maior_cliente = dados["nome"]

print("\n=== MAIOR SALDO ===")
print(f"Cliente: {maior_cliente}")
print(f"Saldo: R$ {maior_saldo:.2f}")
```
