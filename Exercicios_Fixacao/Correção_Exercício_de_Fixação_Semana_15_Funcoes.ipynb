```python
'''
1 - Em um sistema de controle de acesso, é necessário verificar se um
valor ultrapassa um limite definido. Elabore um programa em Python que
solicite ao usuário a entrada de um número inteiro e determine se ele é
maior que 10. Implemente a função verificar_maior_que_dez(numero), que
receba um inteiro e retorne um valor booleano. O programa deve exibir
uma mensagem descritiva ao usuário com base no resultado.'''

# Funções -----------------------------------------------------
def verificar_maior_que_dez(numero):
    if numero>10:
        return True
    else:
        return False

# Programa ----------------------------------------------------
valor = int(input('Digite um número inteiro: '))
if verificar_maior_que_dez(valor):
    print('Maior que 10')
else:
    print('Igual ou Menor que 10')
```

```python
'''
2 - Em sistemas de validação de dados numéricos, identificar se um valor
é par ou ímpar é uma operação fundamental. Desenvolva um programa em
Python que leia um número inteiro fornecido pelo usuário e informe sua
paridade. Crie a função eh_par(numero) que retorne True se o número for
divisível por 2 e False caso contrário. O programa deve exibir o
resultado de forma clara.'''

def eh_par(numero):
    if numero%2 == 0: return True
    else: return False

num = int(input('Digite um valor inteiro:'))
print(eh_par(num))
```

```python
'''
3 - Em um sistema de cadastro, regras de validação de nomes são aplicadas
automaticamente. Escreva um programa que leia um nome informado pelo
usuário e verifique se ele contém a letra "a" (maiúscula ou minúscula).
Implemente a função contem_letra_a(nome) que retorne um valor booleano
conforme a condição. Exiba uma mensagem adequada ao usuário.
'''
def contem_letra_a(nome):
    if ('a' or 'A') in nome: return True
    else: return False

palavra = input('Digite um nome: ')
print(contem_letra_a(palavra))
```

```python
'''
4 - Em sistemas de análise numérica, a divisibilidade é um critério frequentemente utilizado.
Desenvolva um programa que leia um número inteiro e informe se ele é múltiplo de 5.
Crie a função eh_multiplo_de_5(numero) que retorne um valor lógico com base na
divisibilidade por 5. O programa deve exibir o resultado de forma interpretativa.
'''
def eh_multiplo_de_5(numero):
    if numero%5==0: return True
    else: return False

num = int(input('Digite o valor desejado: '))
print(f'O valor é múltiplo de 5? {eh_multiplo_de_5(num)}')
```

```python
'''
5 - Em um sistema de monitoramento ambiental, as temperaturas precisam ser classificadas
para acionar alertas. Escreva um programa que leia a temperatura atual em graus Celsius e
informe se o clima está quente (acima de 30°C) ou frio (30°C ou menos). Implemente a função
classificar_temperatura(temp) que retorne a classificação correspondente como string.
'''
def classificar_temperatura(temp):
    if temp > 30: return 'quente'
    else: return 'frio'


celsius = float(input('Digite a temperatura: '))
print(f'O dia está {classificar_temperatura(celsius)}')
```

```python
'''
    6 - Em sistemas interativos, o usuário frequentemente precisa confirmar ou cancelar ações.
    Desenvolva um programa em Python que solicite ao usuário a entrada de um caractere:
    "S" para continuar ou "N" para sair.
    Crie a função validar_opcao(opcao) que retorne a mensagem adequada para cada caso,
    tratando também entradas inválidas.'''

def validar_opcao(opcao):
    if opcao.upper() == 'S':
        print('Continuando...')
    elif opcao.upper() == 'N':
        print('Saindo...')
    else:
        print('Opção Inválida')

option = input('Digite "S" para continuar ou "N" para sair: ')
validar_opcao(option)


```

```python
''' 7 - Em sistemas de análise financeira e estatística, valores precisam ser categorizados
automaticamente. Escreva um programa que leia um número qualquer e informe se ele é negativo,
zero ou positivo. Implemente a função classificar_numero(numero) que retorne a classificação
como string descritiva.'''

def classificar_numero(numero):
    if numero > 0:
        return 'Positivo'
    elif numero < 0:
        return 'Negativo'
    else:
        return 'Zero'

num = int(input('Digite um valor: '))
print(f'O número {num} é um número {classificar_numero(num)}')
```

```python
''' 8 - Em sistemas de agenda eletrônica, números são frequentemente associados a dias da semana
para automatizar processos de agendamento. Desenvolva um programa em Python que leia um número
de 1 a 7 e retorne o nome do dia correspondente (1 = Segunda-feira, …, 7 = Domingo).
Implemente a função dia_da_semana(numero) tratando entradas fora do intervalo esperado.'''

def dia_da_semana(numero):
    match numero:
        case 1: return 'Segunda-Feira'
        case 2: return 'Terça-Feira'
        case 3: return 'Quarta-Feira'
        case 4: return 'Quinta-Feira'
        case 5: return 'Sexta-Feira'
        case 6: return 'Sábado'
        case 7: return 'Domingo'
        case _: return 'inválido'

dia = int(input('Digite o valor do dia da semana (1 - 7): '))
print(f'O dia da semana é {dia_da_semana(dia)}')
```

```python
''' 9 - Em sistemas de calendário digital, a identificação de anos bissextos é essencial para
 o correto cômputo de datas. Crie um programa que leia um ano e informe se ele é bissexto.
 Implemente a função eh_bissexto(ano) com base na regra: divisível por 4, exceto centenas —
 salvo as divisíveis por 400.'''
def eh_bissexto(ano):
    if (ano%4 == 0) and (not((ano%100==0)) or (ano%400==0)): return 'Ano Bissexto'
    else: return 'Ano não Bissexto'

year = int(input('Digite o ano desejado: '))
print(eh_bissexto(year))

```

```python
''' 10 - Em um sistema de folha de pagamento, o cálculo deve considerar horas normais e
excedentes. Desenvolva um programa que leia a quantidade de horas trabalhadas e o valor
por hora. Até 40h: pagamento normal. Acima de 40h: horas excedentes com acréscimo de 50%.
Implemente a função calcular_salario(horas, valor_hora) que retorne o salário total calculado.'''

def calcular_salario(horas, valor_hora):
    if horas <= 40:  return horas*valor_hora
    else: return (40 * valor_hora) + ((horas-40)*(valor_hora*1.5))

salario = float(input('Digite o valor da hora trabalhada: R$ '))
hora_trabalhada = int(input('Quantas horas foram trabalhadas? '))
print(f'Seu salário total é de R${calcular_salario(hora_trabalhada,salario):.2f}')

```

```python
''' 11 - Em sistemas de análise estatística, é comum agregar valores armazenados em estruturas
lineares. Crie a função soma_vetor(vetor) que receba uma lista de números inteiros e retorne a
soma de todos os seus elementos, sem utilizar a função built-in sum(). Teste com 10 valores
fornecidos pelo usuário.'''
def soma_vetor(vetor):
    soma = 0
    for n in vetor: soma+=n
    return soma

numeros = [1,3,4,5,8,7,9,6,10,2]
print(soma_vetor(numeros))
```

```python
''' 12 - Em um sistema de cadastro corporativo, nomes precisam ser organizados em ordem alfabética
para facilitar a navegação. Implemente a função ordenar_nomes(nomes) que receba uma lista com 5
nomes e retorne uma nova lista ordenada alfabeticamente, sem modificar a lista original.
Leia os nomes via entrada do usuário.'''

def ordenar_nome(nomes):
    nomes.sort() #lista_ordenada = sorted(nomes)
    return nomes #return lista_ordenada

names = []
for nome in range(5):
    name = input('Digite um nome: ')
    names.append(name)
print(ordenar_nome(names))
```

```python
''' 13 - Em análise de dados, determinar os valores extremos de um
conjunto é operação essencial. Crie a função maior_menor(vetor) que
receba uma lista de 8 números inteiros e retorne uma tupla no formato
(maior, menor), sem utilizar as funções built-in max() e min().
'''
def maior_menor(vetor):
    vetor.sort()
    return (vetor[-1],vetor[0])

lista = [8,5,2,3,6,9,7,4]
dados = maior_menor(lista)
print(dados)
```

```python
''' 14 - Em um sistema de processamento de lotes, apenas valores com
determinadas propriedades devem ser retidos. Implemente a função
filtrar_pares(lista) que receba uma lista de inteiros e retorne uma nova
lista contendo apenas os valores pares, preservando a ordem original.
Teste com 10 números fornecidos pelo usuário.'''

def filtrar_pares(lista):
    pares = []
    for item in lista:
        if item%2==0:
            pares.append(item)
    return pares

lista = [8,5,2,3,6,9,7,4,1,10]
print(filtrar_pares(lista))
```

```python
''' 15 - Em estatística aplicada a sistemas computacionais, o cálculo da
média é operação central. Crie a função media_valores(lista) que receba
uma lista de números reais e retorne a média aritmética com duas casas
decimais. A função deve tratar o caso de lista vazia, retornando 0.
Teste com 6 elementos fornecidos pelo usuário. '''
def media_valores(lista):
    if len(lista) == 0:
        return 0
    else:
        soma = 0
        for item in lista: soma+=item
        media = soma/len(lista)
        return media

lista_num = []
for i in range(6):
    num = float(input(f'Digite o {i+1}º valor: '))
    lista_num.append(num)
print(f'{media_valores(lista_num):.2f}')
```

```python
''' 16 - Em análise de desempenho, conta-se quantos valores de um
conjunto superam uma referência. Implemente a função contar_maiores
(lista, valor_referencia) que retorne um inteiro indicando quantos
elementos da lista são estritamente maiores que o valor informado.
Teste com um vetor de 10 elementos e compare com o valor 50.'''

def contar_maiores(lista, valor_referencia):
    cont = 0
    for item in lista:
        if item > valor_referencia:
            cont+=1
    return cont

vetor = [192,3221,323,45,633,32423,54,65,7,6,34,2244,52,52,545,46,346]
print(contar_maiores(vetor,50))
```

```python
''' 17 - Em processamento de dados sequenciais, a inversão da ordem de
elementos é operação recorrente. Crie a função inverter_lista(lista)
que retorne uma nova lista com os elementos na ordem inversa da original,
sem utilizar o método .reverse() nem fatiamento. Teste com um vetor de
7 elementos.
'''
def inverter_lista(lista):
    lista_reversa = []
    for i in range(len(lista)-1,-1,-1):
        lista_reversa.append(lista[i])
    return lista_reversa

lista = [1,2,3,4,5,6,7]
print(inverter_lista(lista))
```

```python
''' 18 - Em processos de limpeza e auditoria de dados, identificar
valores repetidos é fundamental para garantir a integridade das
informações. Implemente a função encontrar_duplicatas(lista) que receba
uma lista de inteiros e retorne uma nova lista contendo apenas os
valores que aparecem mais de uma vez, sem repeti-los na saída. Teste com
uma lista de 5 elementos com possíveis repetições.'''

def encontrar_duplicatas(lista):
    repetidos = []
    for item in lista:
        if lista.count(item) > 1:
            repetidos.append(item)
    return list(set(repetidos))

listinha = [1,1,3,3,4]
print(encontrar_duplicatas(listinha))
```
