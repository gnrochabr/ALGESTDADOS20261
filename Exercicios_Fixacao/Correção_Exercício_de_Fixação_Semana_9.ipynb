```python
# ============================================
# 01 - Crescimento Assintótico
# ============================================
# Dois sistemas A e B crescem de formas diferentes:
# A: crescimento linear (incremento fixo)
# B: crescimento percentual (sobre valor atual)
# Objetivo: descobrir em quantas iterações B ultrapassa A

a = float(input('Digite o valor inicial de A: '))
b = float(input('Digite o valor inicial de B: '))
inc_a = float(input('Crescimento fixo de A por iteracao: '))
taxa_b = float(input('Taxa de crescimento de B (ex: 0.05): '))
limite = int(input('Limite maximo de iteracoes: '))

i = 0

while i < limite and b <= a:
    a = a + inc_a
    b = b + (b * taxa_b)
    i += 1

if b > a:
    print('B ultrapassou A em', i, 'iteracoes')
else:
    print('Nao houve ultrapassagem')
```

```python
# ============================================
# 02 - Decomposição Otimizada de Valor
# ============================================
# Decompor um valor em notas: 100, 50, 20, 10, 5, 2 e 1

valor = int(input('Digite um valor inteiro: '))

n100 = n50 = n20 = n10 = n5 = n2 = n1 = 0

while valor >= 100:
    valor -= 100
    n100 += 1

while valor >= 50:
    valor -= 50
    n50 += 1

while valor >= 20:
    valor -= 20
    n20 += 1

while valor >= 10:
    valor -= 10
    n10 += 1

while valor >= 5:
    valor -= 5
    n5 += 1

while valor >= 2:
    valor -= 2
    n2 += 1

while valor >= 1:
    valor -= 1
    n1 += 1

print('Notas utilizadas:')
print('100:', n100, '| 50:', n50, '| 20:', n20)
print(' 10:', n10,  '|  5:', n5,  '|  2:', n2, '| 1:', n1)
```

```python
# ============================================
# 03 - Sequência Iterativa Não Linear
# ============================================
# Se n for par  → n = n / 2
# Se n for ímpar → n = 5n + 1
# Parar ao atingir 1 ou exceder o limite de passos
# CORREÇÃO: o limite agora controla passos, não o valor de n

n = int(input('Digite o valor inicial: '))
limite = int(input('Limite maximo de passos: '))

passos = 0
maior = n

while n != 1 and passos < limite:
    if n % 2 == 0:
        n = n // 2
    else:
        n = 5 * n + 1

    if n > maior:
        maior = n

    passos += 1

print('Passos:', passos)
print('Maior valor atingido:', maior)
if n == 1:
    print('Sequencia convergiu para 1')
else:
    print('Limite de passos atingido sem convergir')
```

```python
# ============================================
# 04 - Validação com Controle de Estado
# ============================================
# Após 3 erros consecutivos → estado de alerta

minimo = int(input('Valor minimo permitido: '))
maximo = int(input('Valor maximo permitido: '))

erros = 0
valor = int(input('Digite um valor: '))

while valor < minimo or valor > maximo:
    print('Valor invalido!')
    erros += 1

    if erros == 3:
        print('ALERTA: muitas tentativas invalidas')
        confirmacao = int(input('Digite 1 para continuar ou qualquer outro numero para sair: '))

        if confirmacao != 1:
            break

        erros = 0

    valor = int(input('Digite novamente: '))

if minimo <= valor <= maximo:
    print('Valor valido!')
else:
    print('Sistema encerrado')
```

```python
# ============================================
# 05 - Aproximação Iterativa (Raiz Quadrada)
# ============================================

n = float(input('Numero para calcular raiz: '))
erro = float(input('Erro tolerado (ex: 0.0001): '))

x = n
dif = 1.0

while dif > erro:
    novo = (x + n / x) / 2
    dif = abs(novo - x) / novo
    x = novo

print('Raiz aproximada:', x)
```

```python
# ============================================
# 06 - Subsequência Crescente
# ============================================
# CORREÇÃO: adicionado prompt nos input() sem mensagem

num = int(input('Digite o primeiro numero (0 encerra): '))

anterior = num
cont = 1
maior = 1

while num != 0:
    num = int(input('Proximo numero (0 encerra): '))

    if num > anterior:
        cont += 1
    else:
        if cont > maior:
            maior = cont
        cont = 1

    anterior = num

print('Maior sequencia crescente:', maior)
```

```python
# ============================================
# 07 - Conversão de Base
# ============================================

n = int(input('Numero decimal: '))
base = int(input('Base (2 a 9): '))

resultado = 0
pos = 1

while n > 0:
    resto = n % base
    resultado += resto * pos
    pos *= 10
    n = n // base

print('Numero convertido:', resultado)
```

```python
# ============================================
# 08 - Simulação Financeira
# ============================================

capital = float(input('Capital inicial: '))
taxa = float(input('Taxa de crescimento (ex: 0.05): '))
meta = float(input('Meta inicial: '))
limite = int(input('Limite de ciclos: '))

i = 0

while i < limite and capital < meta:
    capital += capital * taxa
    meta += meta * 0.01 * (i + 1)
    i += 1

print('Iteracoes:', i)
print('Capital final: R$', round(capital, 2))
print('Meta final:    R$', round(meta, 2))
if capital >= meta:
    print('Meta atingida!')
else:
    print('Meta nao atingida no limite de ciclos')
```

```python
# ============================================
# 09 - Detecção de Número Abundante
# ============================================
# Abundante → soma dos divisores > número
# Perfeito  → soma = número
# Deficiente → soma < número

n = int(input('Digite um numero inteiro positivo: '))

i = 1
soma = 0

while i < n:
    if n % i == 0:
        soma += i
    i += 1

if soma > n:
    print('Classificacao: Abundante')
elif soma == n:
    print('Classificacao: Perfeito')
else:
    print('Classificacao: Deficiente')

print('Soma dos divisores:', soma)
```

```python
# ============================================
# 10 - Compactação de Sequência Numérica
# ============================================
# CORREÇÃO: adicionado prompt nos input() sem mensagem

num = int(input('Digite o primeiro numero (0 encerra): '))

anterior = num
cont = 1

while num != 0:
    num = int(input('Proximo numero (0 encerra): '))

    if num == anterior:
        cont += 1
    else:
        print('Valor:', anterior, '| Frequencia:', cont)
        cont = 1

    anterior = num
```

```python
# ============================================
# 11 - Série Alternada
# ============================================

tol = float(input('Digite a tolerancia (ex: 0.001): '))

termo = 1.0
soma = 0.0
i = 1

while abs(termo) > tol:
    termo = ((-1) ** i) / i
    soma += termo
    i += 1

print('Quantidade de termos:', i - 1)
print('Soma aproximada:', soma)
```

```python
# ============================================
# 12 - Controle de Acesso com Sessões
# ============================================

ativo = 1

while ativo == 1:
    tentativas = 0
    sucesso = 0

    while tentativas < 3 and sucesso == 0:
        senha = int(input('Digite a senha: '))

        if senha == 123:
            sucesso = 1
        else:
            print('Senha incorreta')
            tentativas += 1

    if sucesso == 1:
        print('Login realizado com sucesso')
    else:
        print('Sessao bloqueada')

    ativo = int(input('Deseja iniciar nova sessao? (1=sim / 0=nao): '))
```

```python
# ============================================
# 13 - Detecção de Padrão Repetitivo
# ============================================

a = int(input('Primeiro valor do padrao: '))
b = int(input('Segundo valor do padrao: '))

cont = 0

while True:
    x = int(input('Digite o 1o valor do par: '))
    y = int(input('Digite o 2o valor do par: '))

    if x == a and y == b:
        cont += 1
    else:
        break

print('Repeticoes do padrao:', cont)
print('Tamanho do bloco: 2')
```

```python
# ============================================
# 14 - Fatorial Controlado
# ============================================

n = int(input('Numero para calcular fatorial: '))
limite = int(input('Limite maximo permitido: '))

fat = 1
i = 1

while i <= n:
    fat = fat * i

    if fat > limite:
        print('Overflow detectado na iteracao:', i)
        break

    i += 1

if fat <= limite:
    print('Fatorial calculado:', fat)
```

```python
# ============================================
# 15 - Simulador de Recursos Dinâmicos
# ============================================

recurso = float(input('Quantidade inicial do recurso: '))
consumo = float(input('Consumo por ciclo: '))
reposicao = float(input('Reposicao por ciclo: '))
limite = float(input('Limite maximo de recurso: '))

ciclos = 0

while recurso > 0 and recurso < limite:
    recurso = recurso - consumo + reposicao
    ciclos += 1

print('Total de ciclos executados:', ciclos)
print('Quantidade final do recurso:', round(recurso, 2))
if recurso <= 0:
    print('Recurso esgotado')
else:
    print('Limite maximo atingido')
```
