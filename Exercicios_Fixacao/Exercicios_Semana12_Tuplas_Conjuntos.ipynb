```python
#1 - Crie uma tupla contendo três coordenadas no formato (x, y). Exiba cada coordenada separadamente, calcule a soma de todos os valores armazenados e informe qual é o maior número presente na tupla.

coordenadas = ( (124556, 832192),  (877837, 654322),  (32343432, 543232))
soma = 0
maior = coordenadas[0][0]
for coord in coordenadas:
    x, y = coord
    print("X =", x)
    print("Y =", y)

    soma += x + y

    if x > maior:
        maior = x
    if y > maior:
        maior = y

print("Soma total:", soma)
print("Maior número:", maior)
```

```python
#2 - Leia cinco nomes de produtos digitados pelo usuário e armazene-os em um conjunto. Ao final, exiba todos os produtos cadastrados e a quantidade total de itens únicos.
conjunto = set()
for i in range(5):
    item = input(f'Digite o {i+1}° produto: ')
    conjunto.add(item)
for item in conjunto:
    print(item)
print(f'Produtos únicos: {len(conjunto)}')
```

```python
#3 - Crie uma tupla com 6 números inteiros. Exiba a quantidade de números pares, a soma dos números ímpares e o menor valor armazenado.
numeros = (1,2,3,4,5,6)
menor = numeros[0]
soma = 0
pares = 0
for numero in numeros:
    #Procura os pares
    if numero%2 == 0:
        pares+=1
    else:
    #faz a soma dos ímpares da Tupla
        soma+=numero
    #Procura o menor valor
    if numero < menor:
        menor = numero

print(f'O menor valor é: {menor}\nA quantidade de pares é: {pares}\nA soma dos valores ímpares da tupla é {soma}.')
```

```python
#4 - Crie dois conjuntos contendo nomes de alunos de duas turmas diferentes. Exiba um novo conjunto contendo todos os alunos sem repetição usando a operação de união.
turmaA = {'Zé Chico','João Franscisco','Joaquim','Maria Bonita'}
turmaB = {'Maria Quitéria','João Maria','Permínio','Aguinaldo'}
turmas = turmaA.union(turmaB)
print(turmas)
```

```python
#5 - Crie uma tupla com nomes de cidades. Leia uma cidade digitada pelo usuário
#e verifique se ela existe na tupla.
#Exiba uma mensagem informando se a cidade foi encontrada ou não.
cidades = ('Vila Pavão','Vila Valério','Nova Venécia', 'Boa Esperança','São Gabriel da Palha')
cidade = input('Digite o nome de uma cidade: ')

if cidade in cidades:
    print(f'A cidade de {cidade} foi encontrada')
else:
    print(f'A cidade de {cidade} não foi encontrada')
```

```python
#6 - Leia uma palavra digitada pelo usuário e armazene em um conjunto todas as vogais diferentes
#encontradas nela. Exiba o conjunto final.
palavra = input('Digite uma palavra: ')
vogais = set()
for letra in palavra:
    if letra.lower() in 'aeiou':
        vogais.add(letra.lower())
print(vogais)
```

```python
#7 - Crie uma tupla com quatro notas de um aluno. Exiba a média das notas, a maior nota e a menor nota. Mostre também todas as notas armazenadas.
notas = (10.0,9.5,8.5,9.0)
soma = 0
maior = max(notas)
menor = min(notas)
for nota in notas:
    soma += nota
media = soma/len(notas)
print(f'Média de Notas: {media}, Maior Nota: {maior}, Menor Nota: {menor}')
```

```python
#8 - Crie dois conjuntos com números inteiros. Exiba apenas os números que aparecem nos dois conjuntos utilizando a operação de interseção.
conjuntoA = {2,4,5,6,7,9}
conjuntoB = {5,7,9}
conjuntos = conjuntoA.intersection(conjuntoB)
print(conjuntos)
```

```python
#9 - Leia uma frase e utilize um conjunto para armazenar apenas as letras
#diferentes digitadas pelo usuário. Ignore espaços. Ao final, exiba a
#quantidade de letras únicas encontradas.

frase = input('Filosofe uma frase: ').lower().replace(" ","")
letrinhas=set()
for letra in frase:
    if letra not in letrinhas:
        letrinhas.add(letra)
print(letrinhas)
```

```python
#10 - Crie uma tupla contendo os nomes dos 12 meses do ano.
#Leia um número de 1 a 12 e exiba o mês correspondente usando
#indexação da tupla.
meses = ('Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro')
numero = int(input('Digite um número de 1 a 12: '))
if 1 <= numero <= 12:
    print(meses[numero-1]) #Índice é a posição do elemento - 1
else:
    print('Mês não válido')
```
