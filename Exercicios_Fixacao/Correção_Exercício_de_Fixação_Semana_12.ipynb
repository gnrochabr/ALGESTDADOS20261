```python
#1 - Crie uma tupla contendo três coordenadas no formato (x, y). Exiba cada coordenada separadamente, calcule a soma de todos os valores armazenados e informe qual é o maior número presente na tupla.

coordenadas = ( (124556, 832192),  (877837, 654322),  (32343432, 543232))
soma = 0
maior = coordenadas[0][0]
for coord in coordenadas:
    x, y = coord
    print("X =", x)
    print("Y =", y)

    soma += x + y

    if x > maior:
        maior = x
    if y > maior:
        maior = y

print("Soma total:", soma)
print("Maior número:", maior)
```

```python
#2 - Leia cinco nomes de produtos digitados pelo usuário e armazene-os em um conjunto. Ao final, exiba todos os produtos cadastrados e a quantidade total de itens únicos.
conjunto = set()
for i in range(5):
    item = input(f'Digite o {i+1}° produto: ')
    conjunto.add(item)
for item in conjunto:
    print(item)
print(f'Produtos únicos: {len(conjunto)}')
```

```python
#3 - Crie uma tupla com 6 números inteiros. Exiba a quantidade de números pares, a soma dos números ímpares e o menor valor armazenado.
numeros = (1,2,3,4,5,6)
menor = numeros[0]
soma = 0
pares = 0
for numero in numeros:
    #Procura os pares
    if numero%2 == 0:
        pares+=1
    else:
    #faz a soma dos ímpares da Tupla
        soma+=numero
    #Procura o menor valor
    if numero < menor:
        menor = numero

print(f'O menor valor é: {menor}\nA quantidade de pares é: {pares}\nA soma dos valores ímpares da tupla é {soma}.')
```

```python
#4 - Crie dois conjuntos contendo nomes de alunos de duas turmas diferentes. Exiba um novo conjunto contendo todos os alunos sem repetição usando a operação de união.
turmaA = {'Zé Chico','João Franscisco','Joaquim','Maria Bonita'}
turmaB = {'Maria Quitéria','João Maria','Permínio','Aguinaldo'}
turmas = turmaA.union(turmaB)
print(turmas)
```

```python
#6 - Leia uma palavra digitada pelo usuário e armazene em um conjunto todas as vogais diferentes
#encontradas nela. Exiba o conjunto final.
palavra = input('Digite uma palavra: ')
vogais = set()
for letra in palavra:
    if letra.lower() in 'aeiou':
        vogais.add(letra.lower())
print(vogais)
```

```python
#7 - Crie uma tupla com quatro notas de um aluno. Exiba a média das notas, a maior nota e a menor nota. Mostre também todas as notas armazenadas.
notas = (10.0,9.5,8.5,9.0)
soma = 0
maior = max(notas)
menor = min(notas)
for nota in notas:
    soma += nota
media = soma/len(notas)
print(f'Média de Notas: {media}, Maior Nota: {maior}, Menor Nota: {menor}')
```

```python
#8 - Crie dois conjuntos com números inteiros. Exiba apenas os números que aparecem nos dois conjuntos utilizando a operação de interseção.
conjuntoA = {2,4,5,6,7,9}
conjuntoB = {5,7,9}
conjuntos = conjuntoA.intersection(conjuntoB)
print(conjuntos)
```

```python
#9 - Leia uma frase e utilize um conjunto para armazenar apenas as letras
#diferentes digitadas pelo usuário. Ignore espaços. Ao final, exiba a
#quantidade de letras únicas encontradas.

frase = input('Filosofe uma frase: ').lower().replace(" ","")
letrinhas=set()
for letra in frase:
    if letra not in letrinhas:
        letrinhas.add(letra)
print(letrinhas)
```

```python
#10 - Crie uma tupla contendo os nomes dos 12 meses do ano.
#Leia um número de 1 a 12 e exiba o mês correspondente usando
#indexação da tupla.
meses = ('Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro')
numero = int(input('Digite um número de 1 a 12: '))
if 1 <= numero <= 12:
    print(meses[numero-1]) #Índice é a posição do elemento - 1
else:
    print('Mês não válido')
```

```python
#11 - Crie dois conjuntos representando alunos presentes em dois dias diferentes. Exiba quais alunos compareceram apenas no primeiro dia utilizando a operação de diferença.
primeiro_dia = {'Maria Quitéria','João Maria','Zé Chico','João Franscisco','Joaquim','Maria Bonita'}
segundo_dia = {'Maria Quitéria','João Maria','João Franscisco','Joaquim','Permínio','Aguinaldo'}
print(f'Compareceram no primeiro dia: {primeiro_dia.difference(segundo_dia)}')
```

```python
#12 - Crie uma tupla com temperaturas registradas durante a semana. Exiba quantas temperaturas estão acima da média e informe a maior e a menor temperatura registrada.
temperaturas = (35.6,31.4,25.6,21.4,23.6,26.4,27.5)
media = sum(temperaturas) / len(temperaturas)
maior = max(temperaturas)
menor = min(temperaturas)
print(f'Temperatura Média: {media:.1f}°C\nTemperatura Máxima: {maior}°C\nTemperatura Mínima: {menor}°C')
```

```python
#13 - Leia 10 números inteiros digitados pelo usuário e armazene-os em um conjunto. Ao final, exiba os números digitados sem repetição e informe quantos valores diferentes foram inseridos.
conjunto = set()
for i in range(10):
    valor = int(input(f'Digite o {i+1}° valor:'))
    conjunto.add(valor)
print(f'Os valores do Conjunto são: {conjunto}\nValores Diferentes: {len(conjunto)}')
```

```python
#14 - Crie uma tupla com 8 números inteiros. Leia um número e verifique em quais posições ele aparece dentro da tupla. Caso não exista, exiba uma mensagem apropriada.
numeros = (2,2,3,3,2,4,5,6)
verificador = int(input('Digite sua busca: '))
if verificador not in numeros:
    print(f'O número {verificador} não está na Tupla.')
else:
    for posicao in range(len(numeros)):
        if verificador ==  numeros[posicao]:
            print(f'O número {verificador} está na Posição: {posicao}')
```

```python
#17 - Crie dois conjuntos de números inteiros. Verifique se os conjuntos são iguais, diferentes ou se um está contido dentro do outro. Exiba mensagens apropriadas para cada caso.
conjuntoA = {2,4,5,6,7,9}
conjuntoB = {5,7,9}
if conjuntoA == conjuntoB:
    print('Os conjuntos são iguais')
else:
    print('Os conjuntos são diferentes')

if conjuntoB.issubset(conjuntoA):
    print('O conjunto B está contido em A')
else:
    print('O conjunto B não está contido em A')

if conjuntoA.issubset(conjuntoB):
    print('O conjunto A está contido em B')
else:
    print('O conjunto A não está contido em B')
```

```python
#20 - Crie uma tupla com 10 números inteiros. Exiba a soma total, a média dos valores, o maior número, o menor número e quantos valores são pares.
num_int = (951,242,578,154,147,124,547,211,21,10)
pares = 0
for num in num_int:
    if num%2 == 0:
        pares+=1
maior = max(num_int)
menor = min(num_int)
soma = sum(num_int)
media = soma/len(num_int)

print(f'Soma Total: {soma}\nMédia: {media}\nMaior valor: {maior}\nMenor valor:{menor}\nPares: {pares}')
```
