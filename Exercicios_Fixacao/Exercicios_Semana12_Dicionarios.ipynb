```python
#1 - Crie um dicionário chamado produto com as chaves: nome, preco, estoque e categoria.
#Imprima cada campo com for e .items().
#Depois, atualize o preço e adicione uma chave desconto com um valor de sua escolha.

produto = {
    'nome':'Arroz Sepé 5kg',
    'preco':14.98,
    'estoque': 200,
    'categoria': 'Perecíveis'
}
for descricao in produto:
    print(f'{descricao}: {produto[descricao]}')

produto['preco'] = 10.90 #Alterando valores
produto['desconto'] = 0.8 #Adicionando chave-valor (20% de desconto)

print(produto.items())
```

```python
# 2 - Crie uma lista onde cada item é um dicionário representando um aluno (com nome e media).
#Use um for para percorrer a lista e imprimir: 'Aluno: [nome] — Média: [media]'.
#Depois, imprima apenas os alunos com média acima de 7.

aluno1 = {'nome': 'Arbileu', 'media': 6.5}
aluno2 = {'nome': 'Birineu', 'media': 5.0}
aluno3 = {'nome': 'Cartileu', 'media': 9.5}
aluno4 = {'nome': 'Dobileu', 'media': 6.5}
aluno5 = {'nome': 'Erbileu', 'media': 4.0}

alunos = [aluno1,aluno2,aluno3,aluno4,aluno5]

#imprimir usando posição
for aluno in alunos:
    print(f'Aluno: {aluno['nome']} - Média: {aluno['media']}')

print('\nAlunos acima de média 7.0:')

for aluno in alunos:
    if aluno['media']>=7:
        print(f'Aluno: {aluno['nome']} - Média: {aluno['media']}')
```

```python
#3 - Crie um dicionário chamado produto com as chaves nome, preco e quantidade; em seguida,
#calcule o valor total em estoque multiplicando o preço pela quantidade e armazene esse resultado
#em uma nova chave chamada valor_total; utilize um for com .items() para exibir todas as
#informações do produto, incluindo o valor total calculado.

produto = {
    'nome':'Arroz Sepé 5kg',
    'preco':14.98,
    'quantidade': 200,
}

valor_total = produto['preco']*produto['quantidade']

#Imprimir usando .items()
for descricao, valor in produto.items():
    print(f'{descricao} - {valor}')

print(f'Valor Total: R$ {valor_total:.2f}')
```
