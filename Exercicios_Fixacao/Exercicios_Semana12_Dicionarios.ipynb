```python
# 5 - Crie um dicionário com informações de um livro (titulo, autor, ano).
#Escreva um programa que, para cada uma das chaves titulo, editora e
#paginas, verifique se ela existe no dicionário e imprima uma mensagem
#informando se está presente ou ausente. Use o operador in para as
#verificações.
livro = {'titulo':'Dom Quixote', 'autor':'Miguel de Cervantes', 'ano':1605}
if livro.get('titulo') in livro.values():
    print('Ok - Tem título')
else:
    print('Não tem título')

if livro.get('autor') in livro.values():
    print('Ok - Tem autor')
else:
    print('Não tem autor')

if livro.get('ano') in livro.values():
    print('Ok - Tem ano')
else:
    print('Não tem ano')

if livro.get('página') in livro.values():
    print('Ok - Tem página')
else:
    print('Não tem páginas')
```

```python
#6 - Dado um dicionário de agenda telefônica com pelo menos 5 contatos
#(onde cada chave é um nome e o valor é um número de telefone), exiba:
#(a) quantos contatos há na agenda usando len();
#(b) todos os nomes usando .keys();
#(c) todos os telefones usando .values().
agenda = {
    'Fulano':'27999858585',
    'Cicrano':'27999747474',
    'Beltrano':'27999696969',
    'Lorem':'27999313131',
    'Ipsum':'27999272727'
}
print('a) Contatos na agenda:', len(agenda))
print('b) Nomes na agenda: ',end=' ')
for nome in agenda.keys():
    print(', '.join(agenda.keys()))
print('\nc) Valores na agenda: ',end=' ')
for nome in agenda.values():
    print(nome, end = ' ')
```

```python
#7 - Crie um dicionário chamado notas com o nome de 5 disciplinas como chaves e as notas finais
# (de 0 a 10) como valores. Use um laço for com o método .items() para imprimir cada disciplina
# e sua respectiva nota no formato: "Matemática: 8.5".

notas = {
    'matemática' : 8.5,
    'geografia': 9.0,
    'lingua portuguesa':7.0,
    'historia': 9.0,
    'ciencias': 8.0
}

for disciplina, nota in notas.items():
    print(f'{disciplina.capitalize()}: {nota}')
```

```python
#8 - Dada a string "o gato viu o rato e o rato correu do gato", escreva um programa que conte
#quantas vezes cada palavra aparece e armazene o resultado em um dicionário.
#Trate tudo em letras minúsculas. Exiba os pares palavra–frequência ao final.
#Use .split() para separar as palavras.


frase = "o gato viu o rato e o rato correu do gato" #frase é uma lista de caracteres
lista_palavras = frase.split(' ') #cria a lista de palavras
conjunto_palavras = sorted(set(lista_palavras)) #lista ordenada em conjunto
dicionario_palavras = {} #cria o dicionário vazio

for palavra in conjunto_palavras:
    contador = 0
    for ocorrencia in lista_palavras:
        if palavra == ocorrencia:
            contador+=1
    dicionario_palavras[palavra] = contador

for palavra, quantidade in dicionario_palavras.items():
    print(f'{palavra}: {quantidade}')
```

```python
#9 - Dado o dicionário
#capitais = {"Brasil": "Brasília", "França": "Paris", "Japão": "Tóquio", "Egito": "Cairo"},
#crie um novo dicionário invertido onde as capitais tornam-se chaves e os países tornam-se valores.
#Exiba o dicionário original e o invertido.

capitais = {"Brasil": "Brasília", "França": "Paris", "Japão": "Tóquio", "Egito": "Cairo"}
paises = {}
for pais, capital in capitais.items():
    paises[capital] = pais
print(paises.items())
```

```python
#10 - Você tem duas listas: alunos = ["Ana", "Bruno", "Carla", "Diego"]
#e notas = [9.0, 7.5, 8.2, 6.8]. Combine-as em um único dicionário usando a função zip(),
#onde os nomes são as chaves e as notas são os valores. Exiba o dicionário resultante.

alunos = ["Ana", "Bruno", "Carla", "Diego"]
notas = [9.0, 7.5, 8.2, 6.8]
dicionario = dict(zip(tuple(alunos),tuple(notas)))
print(dicionario)
```
