```python
numeros = (1,2,3,2,4,2,5)

#Index busca o primeiro índice compatível com o valor procurado
indice = numeros.index(2)
print(f'O primeiro número 2 aparece no índice {indice}')

#Count foi feito para contar as vezes que ele aparece
contador = numeros.count(2)
print(f'O número 2 aparece {contador} vezes')
```
